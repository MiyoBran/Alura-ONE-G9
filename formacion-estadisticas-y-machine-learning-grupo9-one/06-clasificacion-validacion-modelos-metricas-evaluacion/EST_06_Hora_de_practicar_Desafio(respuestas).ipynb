## 1️⃣ Intervalo de confianza en validación cruzada

En el proceso de **validación cruzada**, se generan diferentes modelos para cada división realizada en los datos y, en consecuencia, **diferentes valores de métricas de evaluación**.

Para encontrar un resultado promedio de las métricas, se puede **construir un intervalo de confianza** a partir de la **media** y la **desviación estándar** de las métricas.

### Tarea

La función debe recibir **2 parámetros**:

- Una **lista con los resultados de las métricas** obtenidas en la validación cruzada.
- El **nombre del algoritmo** evaluado.

### Pasos a seguir

1. Calcular la **media** de los resultados de la lista.
2. Calcular la **desviación estándar**.
3. Construir el **intervalo de confianza**:

\[
\text{Intervalo} =
\left[

\text{media} + 2 \times \text{desviación estándar}
\right]
\]

# 4️⃣ Validación cruzada con LeaveOneOut

En **conjuntos de datos pequeños**, las estrategias tradicionales de separación pueden dejar **muy pocos datos para entrenamiento**, dificultando que el modelo aprenda el patrón de los datos.

**LeaveOneOut** es una estrategia diseñada para resolver este problema.

### Funcionamiento

- Se utiliza **solo un registro para validación**.
- **Todos los demás registros** se utilizan para entrenamiento.
- El proceso se repite **una vez por cada fila del dataset**.

Esto significa que si el dataset tiene:

```

1000 filas → se entrenan 1000 modelos

```

### Nota importante

El método **LeaveOneOut** generará un modelo para **cada fila del dataset**.

Por lo tanto, la lista de resultados tendrá **solo valores de acierto 0 o 1** para cada modelo.

### Desafío 1

Crea una función para calcular el intervalo de confianza de los resultados de una validación cruzada con 2 desviaciones estándar. La función debe tener un parámetro para recibir una lista con los desempeños de la validación cruzada y otro para presentar el nombre del modelo utilizado en la validación cruzada.  

Ejemplo de retorno de la función:  

> Intervalo de confianza ("nombre del modelo"): ["valor mínimo del intervalo", "valor máximo del intervalo"]


`````python
def intervalo_conf(resultados, nombre_modelo):
    media = resultados.mean()
    desvio_tipico = resultados.std()
    print(f'Intervalo de confianza ({nombre_modelo}): [{media - 2*desvio_tipico}, {min(media + 2*desvio_tipico, 1)}]')
`````
