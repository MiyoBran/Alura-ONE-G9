Este bloque aplica **Estandarizaci√≥n de Esquema** (Snake Case) y **An√°lisis Sem√°ntico Automatizado** para evitar el uso manual de `print` y `f-strings`.


```python

# 1. Ingesta y Estandarizaci√≥n de Nombres (Clean Code)
# Reemplazamos espacios por "_" y convertimos a min√∫sculas para habilitar el acceso por punto.
url = "https://raw.githubusercontent.com/alura-es-cursos/Curso-de-Estadistica-con-Python/Aula-1/datos.csv"
datos = pd.read_csv(url)

# Diccionario de mapeo para profesionalizar el dataset
datos.columns = [col.lower().replace(" ", "_") for col in datos.columns]

# 2. An√°lisis de Rango Autom√°tico (Evita Magic Numbers y redundancia)
def reporte_rango_variable(df, columna):
    """
    Genera un reporte de l√≠mites para una variable cuantitativa.
    Basado en l√≥gica de negocio para control de inventarios/precios.
    """
    min_val = df[columna].min()
    max_val = df[columna].max()
    print(f"[REPORTE {columna.upper()}] M√≠n: {min_val} | M√°x: {max_val}")

# 3. Identificaci√≥n Sem√°ntica (Filtro por Tipo de Dato)
# En producci√≥n, agrupamos variables por su naturaleza t√©cnica.
vars_cuantitativas = ['edad', 'altura', 'ingreso']
vars_cualitativas = ['sexo', 'color', 'a√±os_de_estudio']

# Ejecuci√≥n automatizada (Scalable Data Science)
print("--- RESUMEN DE VARIABLES CUANTITATIVAS ---")
for var in vars_cuantitativas:
    reporte_rango_variable(datos, var)

print("\n--- CATEGOR√çAS √öNICAS (CUALITATIVAS) ---")
# Usamos List Comprehension para un c√≥digo m√°s limpio y r√°pido
categorias = {col: sorted(datos[col].unique()) for col in vars_cualitativas}
for col, values in categorias.items():
    print(f"Variable: {col} | Valores: {values}")

# Vista previa profesional
datos.sample(5)
```

### üìù Apunte: Distribuciones de Frecuencia Cualitativa

La **Distribuci√≥n de Frecuencia** es una herramienta para resumir datos categ√≥ricos. Nos dice cu√°ntas veces (frecuencia absoluta) o qu√© proporci√≥n del total (frecuencia relativa) pertenece a cada categor√≠a.

1. **An√°lisis Univariado (`value_counts`):** Mira una sola variable (Ej: ¬øCu√°ntos hombres y mujeres hay?).
2. **An√°lisis Bivariado (`pd.crosstab`):** Cruza dos variables (Ej: ¬øCu√°l es el ingreso promedio de las mujeres en Bogot√° vs. Medell√≠n?). Es vital para identificar sesgos o nichos de mercado.

---

### üíª Refactorizaci√≥n Senior (Clean Code)

Para ser productivos, vamos a evitar renombrar filas una por una. Usaremos el m√©todo `.map()` y **Method Chaining** para que el c√≥digo sea una sola tuber√≠a limpia de procesamiento.


```python
# Diccionarios de mapeo (Definidos una sola vez)
MAP_SEXO = {0: 'Masculino', 1: 'Femenino'}
MAP_COLOR = {0: 'Indio', 2: 'Blanco', 4: 'Negro', 6: 'Amarillo', 8: 'Moreno', 9: 'Sin declarar'}

# --- 1. FRECUENCIA UNIVARIADA (SEXO) ---
# Usamos concatenaci√≥n para crear la tabla de un solo golpe
frecuencia = datos['sexo'].value_counts()
porcentaje = datos['sexo'].value_counts(normalize=True) * 100

dist_frec_sexo = pd.DataFrame({'Frecuencia': frecuencia, 'Porcentaje (%)': porcentaje})
dist_frec_sexo.index = dist_frec_sexo.index.map(MAP_SEXO)
dist_frec_sexo.rename_axis('Sexo', axis='columns', inplace=True)

print("Distribuci√≥n por Sexo:")
print(dist_frec_sexo.round(2))

# --- 2. AN√ÅLISIS BIVARIADO (CROSSTAB SENIOR) ---
# El reto del l√≠der: Proporci√≥n de ingresos por Sexo y Color
# En lugar de renombrar DESPU√âS, mapeamos ANTES de cruzar (m√°s eficiente)

tabla_ingresos = pd.crosstab(
    index = datos['sexo'].map(MAP_SEXO),
    columns = datos['color'].map(MAP_COLOR),
    values = datos['ingreso'],
    aggfunc = 'mean'
).round(2)

print("\nIngreso Promedio por Sexo y Color (Miles de Pesos):")
print(tabla_ingresos)
```


### ¬øPor qu√© este enfoque es "Senior"?

1. **Uso de `.map()`:** En lugar de crear la tabla con n√∫meros (0, 1) y luego renombrar los ejes con `rename`, transformamos los datos al vuelo. Esto evita errores si el √≠ndice cambia de posici√≥n.
2. **Sin `inplace=True`:** En producci√≥n, preferimos reasignar la variable o encadenar m√©todos. Esto hace que el c√≥digo sea m√°s predecible y f√°cil de depurar.
3. **Encapsulamiento en Crosstab:** Pasamos las funciones de agregaci√≥n (`aggfunc='mean'`) directamente. Esto es lo que un Arquitecto de Datos usa para generar reportes r√°pidos de KPIs (Key Performance Indicators).

### El Desaf√≠o del L√≠der (Tu turno)

La instructora te dej√≥ un ejercicio: calcular la **proporci√≥n** (porcentaje) dentro del `crosstab`. En Pandas, esto se logra con el par√°metro `normalize`.


```python
# margins=True: This argument adds row and column sums (margins) to the table,
# showing the total frequencies for each 'sexo' and 'color' category, as well as
# the grand total.
# normalize=True: This argument converts the frequencies into proportions
# (percentages if multiplied by 100) of the total, making it easier to see the
# relative distribution of the categories.
frecuencia = pd.crosstab(datos['sexo'], datos['color'],\
                         normalize=True , margins = True )
frecuencia.rename(index=sexo,inplace=True)
frecuencia.rename(columns=color,inplace=True)
display(frecuencia.round(4)*100)
```

Excelente, Miyen. Esta parte es clave para tu perfil gerencial: el **An√°lisis Bivariado** es lo que permite cruzar variables para encontrar disparidades, nichos o comportamientos espec√≠ficos que un promedio general siempre oculta.

Aqu√≠ tienes el apunte t√©cnico y la resoluci√≥n del "Desaf√≠o del L√≠der" con est√°ndares de producci√≥n.

---

### üìù Apunte: An√°lisis Bivariado con `pd.crosstab`

El an√°lisis bivariado busca entender la relaci√≥n entre dos variables. Cuando ambas son cualitativas (o usamos una cuantitativa para agregar), la herramienta reina es la **Tabla de Contingencia**.

* **¬øPara qu√© sirve?** En el sector comercial, sirve para responder: "¬øCu√°l es el ticket promedio de los hombres en la sucursal A vs. las mujeres en la sucursal B?".
* **Componentes Clave:**
* **√çndice e Intersecci√≥n:** Define las filas y columnas del cruce.
* **Valores y Agregaci√≥n:** Si queremos analizar una tercera variable num√©rica (como `ingreso`), usamos `values` y una funci√≥n (`aggfunc`), como la media (`mean`) o la suma (`sum`).



---

### üíª Refactorizaci√≥n Senior: El Desaf√≠o de la Proporci√≥n

Como Arquitecto de Datos, vamos a resolver el pedido de tu l√≠der: **¬øCu√°l es el porcentaje de distribuci√≥n de etnias seg√∫n el sexo?** Para que este c√≥digo sea de "producci√≥n", usaremos el par√°metro `normalize`. Esto evita hacer c√°lculos manuales y reduce el riesgo de errores en el reporte.



```python
# Mapeos profesionales (mantenemos los diccionarios definidos)
sexo = {0: 'Masculino', 1: 'Femenino'}
color = {0: 'Indio', 2: 'Blanco', 4: 'Negro', 6: 'Amarillo', 8: 'Moreno', 9: 'Sin declarar'}

# CAMINO: Normalizaci√≥n por FILAS (index)
# Esto responde: "Del 100% de hombres, ¬øqu√© % es de cada etnia?"
dist_porcentual_etnia_sexo = pd.crosstab(
    index = datos['sexo'].map(sexo),
    columns = datos['color'].map(color),
    normalize = 'index' # 'index' normaliza por filas, 'columns' por columnas
) * 100

# Presentaci√≥n con formato profesional (Method Chaining)
reporte_final = dist_porcentual_etnia_sexo.round(2).rename_axis('Distribuci√≥n por Color (%)', axis=1)

print("Reporte de Proporci√≥n √âtnica por Sexo:")
print(reporte_final)

```

### Diagn√≥stico de la Mejora:

1. **Par√°metro `normalize='index'`:** Es mucho m√°s robusto que dividir manualmente. Pandas se encarga de que la suma horizontal de cada fila sea exactamente 1.0 (o 100%).
2. **Mapeo Predictivo:** Al usar `.map(sexo)` dentro del `crosstab`, eliminamos la necesidad de usar `rename` e `inplace=True` despu√©s. El DataFrame nace ya con los nombres correctos.
3. **Sem√°ntica de Negocio:** Renombramos el eje de las columnas a "Distribuci√≥n por Color (%)" para que el l√≠der entienda inmediatamente qu√© unidad de medida est√° viendo sin preguntar.

## <font color=#56B4E9>2.3 Distribuci√≥n de frecuencia para variables cuantitativas (clases de amplitud fija)</font>
***

Sturges: como definir segun nuestra cantidad de clases

### Importando bibliotecas

http://www.numpy.org/

```python
import numpy as np
```

### Paso 1 - Difiniendo el n√∫mero de clases

#### Regla de Sturges

# $$k = 1 + \frac {10}{3}\log_{10}n$$

```python
# El tama√±o de nuestra base de datos es n
#con datos.shape obtenemos la cantidad de (registros , columnas)
display(datos.shape)
#n=registros
n = datos.shape[0]
display(n)
```

```python
#Ahora calculamos k
k = 1 + (10/3)*np.log10(n)
display(k)
```

### Paso 2 - Crear la tabla de frecuencias

```python
# Con sturges obtenemos como recomendacion hacer 17 clases ( a diferencia de las
# 5 personalizadas que teniamos)
frecuencia_sturges = pd.cut(datos.ingreso, bins=int(k), include_lowest=True).value_counts(sort=False)
frecuencia_sturges
```

### El Procedimiento Senior

En lugar de presentar el dato en crudo, separaremos la **capa de c√°lculo** de la **capa de presentaci√≥n**.

1. Calcularemos los cortes una sola vez (para no duplicar el uso de memoria).
2. Construiremos el DataFrame unificando frecuencia absoluta y relativa.
3. Crearemos una funci√≥n de formateo (Mapping) que intercepte esos intervalos de Pandas y los traduzca a rangos comerciales legibles, forzando a que el piso visual sea 0.

Aqu√≠ tienes la refactorizaci√≥n lista para producci√≥n:

---


```python
# 1. C√°lculo de Par√°metros (Regla de Sturges)
n = datos.shape[0]
k = int(1 + (10/3) * np.log10(n))

# 2. Partici√≥n de Datos (Single Source of Truth)
# Cortamos la serie una sola vez. Esto es crucial en Big Data para ahorrar CPU.
cortes_sturges = pd.cut(datos['ingreso'], bins=k, include_lowest=True)

# 3. Agregaci√≥n (Method Chaining)
# Calculamos frecuencias y porcentajes sobre la serie ya cortada
frecuencia = cortes_sturges.value_counts(sort=False)
porcentaje = cortes_sturges.value_counts(normalize=True, sort=False) * 100

# 4. Ensamblaje del DataFrame
tabla_sturges = pd.DataFrame({
    'Frecuencia': frecuencia,
    'Porcentaje (%)': porcentaje.round(3)
})

# 5. Capa de Presentaci√≥n (Clean Visuals)
# Extraemos los l√≠mites de los objetos 'Interval' de Pandas para darles formato de negocio
def formatear_rango(intervalo):
    # max(0, ...) elimina el piso negativo artificial creado por pd.cut
    limite_inf = max(0, intervalo.left)
    limite_sup = intervalo.right
    # Formateamos con separador de miles para facilitar la lectura financiera
    return f"$ {limite_inf:,.0f} a $ {limite_sup:,.0f}"

# Aplicamos el formato al √≠ndice y renombramos sin usar inplace=True
tabla_sturges.index = tabla_sturges.index.map(formatear_rango)
tabla_sturges = tabla_sturges.rename_axis('Rango de Ingreso (Sturges)', axis='index')

# Resultado listo para exportar
print(tabla_sturges)
```

$k = 1 + \frac{10}{3}\log_{10}(n)$

**¬øC√≥mo lo aplicamos a la L√≥gica de Negocio?**
Imagina que est√°s armando el dep√≥sito principal. Tienes miles de productos de distintos tama√±os ($n$), pero a√∫n no sabes cu√°ntas estanter√≠as ($k$) necesitas comprar.

* Si compras muy pocas estanter√≠as (pocas clases), mezclar√°s tornillos con motores y perder√°s el detalle.
* Si compras demasiadas (muchas clases), tendr√°s estanter√≠as vac√≠as y el dep√≥sito ser√° ineficiente.

Sturges asume que tus datos tienen un comportamiento normal (Campana de Gauss) y te da el n√∫mero de estanter√≠as "perfecto" para que la distribuci√≥n de tus datos se revele de forma clara y sin "ruido". Es el est√°ndar de oro para el **An√°lisis Exploratorio de Datos (EDA)** cuando te enfrentas a una variable num√©rica nueva y no tienes reglas de negocio predefinidas.

## <font color=#56B4E9>3.2 Mediana</font>
***

Para obtener la mediana de un conjunto de datos, debemos proceder de la siguiente manera:
1. Ordenar el conjunto de datos;
2. Identificar el n√∫mero de observaciones (registros) del conjunto de datos ($n$);
3. Identicar el elemento del medio:

> Cuando $n$ sea impar, la posici√≥n del elemento del medio se obtendr√° de la siguiente manera:


# $$Elemento_{Md} = \frac{n+1}{2}$$

> Cuando $n$ sea par, la posici√≥n del elemento mediano se obtendr√° de la siguiente manera:


# $$Elemento_{Md} = \frac{n}2$$

4. Obtener la mediana:

> Cuando $n$ sea impar:


# $$Md = X_{Elemento_{Md}}$$

> Cuando $n$ sea par:


# $$Md = \frac{X_{Elemento_{Md}} + X_{Elemento_{Md}+1}}2$$
***

### üìù Apunte Te√≥rico: Medidas de Tendencia Central

Las medidas de tendencia central buscan resumir en un solo n√∫mero el "centro de gravedad" de nuestros datos. En la gesti√≥n de una PyME, nos ayudan a entender el comportamiento del cliente, pero debemos elegir la m√©trica correcta para no tomar decisiones equivocadas.

#### 1. Media Arem√©tica (Promedio)

Es el centro matem√°tico o punto de equilibrio. Se calcula sumando todos los valores y dividiendo por la cantidad de observaciones.

* **Poblaci√≥n:** $\mu = \frac{1}{N}\sum_{i=1}^{N}X_i$
* **Muestra:** $\bar{X} = \frac{1}{n}\sum_{i=1}^{n}X_i$
* **Riesgo de Negocio:** Es extremadamente sensible a valores at√≠picos (*outliers*). Si 9 clientes compran `$1.000` y 1 cliente VIP compra `$100.000`, el promedio dir√° que cada cliente gasta `~$10.900`, lo cual falsea la realidad comercial.

#### 2. Mediana ($Md$)

Es el valor que divide a los datos ordenados exactamente en dos mitades iguales (el percentil 50). Representa al "cliente t√≠pico".

* **C√°lculo (n impar):** $Md = X_{\frac{n+1}{2}}$
* **C√°lculo (n par):** $Md = \frac{X_{\frac{n}{2}} + X_{\frac{n}{2}+1}}{2}$
* **Ventaja de Negocio:** Es **robusta**. El cliente VIP de `$100.000` no mover√° la mediana, d√°ndonos una visi√≥n realista del ticket promedio de la base.

### üíª Laboratorio Estad√≠stico & Optimizaci√≥n Big Data (Clean Code)

En este bloque, refactorizamos el c√°lculo manual eliminando el `reset_index` (que genera columnas basura) usando el par√°metro `ignore_index=True`. Luego, aplicamos la v√≠a "Big Data" para analizar la base de datos completa.

```python
# ==============================================================================
# 1. ENFOQUE UNIVERSITARIO (LABORATORIO ESTAD√çSTICO)
# Funci√≥n Clean Code y modularizada para calcular la mediana manualmente
# ==============================================================================

def calcular_mediana_academica(serie: pd.Series) -> float:
    """
    Calcula la mediana paso a paso siguiendo f√≥rmulas matem√°ticas.
    Previene errores de √≠ndices y datos vac√≠os.
    """
    if serie.empty:
        raise ValueError("Error: La serie de datos est√° vac√≠a.")

    # Clean Code: sort_values con ignore_index=True nos ahorra hacer reset_index()
    # y evita arrastrar strings o √≠ndices viejos.
    serie_ordenada = serie.sort_values(ignore_index=True)
    n = len(serie_ordenada)

    # Control de flujo l√≥gico (Par vs Impar)
    if n % 2 != 0:
        # n es impar (Ej: 7 elementos). Divisi√≥n entera (//) nos da el √≠ndice exacto
        indice_md = n // 2
        mediana = serie_ordenada.iloc[indice_md]
    else:
        # n es par (Ej: 6 elementos). Promediamos los dos centrales
        indice_izq = (n // 2) - 1
        indice_der = n // 2
        mediana = (serie_ordenada.iloc[indice_izq] + serie_ordenada.iloc[indice_der]) / 2

    return float(mediana)

# Prueba del Laboratorio:
mediana_maria = calcular_mediana_academica(df['Mar√≠a'])
mediana_pedro = calcular_mediana_academica(df['Pedro'].sample(6, random_state=200))

print(f"LABORATORIO -> Mediana Mar√≠a (Impar): {mediana_maria}")
print(f"LABORATORIO -> Mediana Pedro (Par): {mediana_pedro}")


# ==============================================================================
# 2. ENFOQUE PRODUCCI√ìN (OPTIMIZACI√ìN BIG DATA)
# ==============================================================================
# En el mundo real, no iteramos ni buscamos √≠ndices a mano.
# Delegamos el c√°lculo a la capa C de Pandas (vectorizaci√≥n) mediante .agg()

# Mapeo preventivo de la variable categ√≥rica para lectura gerencial
map_sexo = {0: 'Masculino', 1: 'Femenino'}

# Method Chaining + Agregaci√≥n M√∫ltiple (.agg)
reporte_tendencia = (
    datos.assign(sexo_desc = datos['sexo'].map(map_sexo)) # Creamos columna temporal legible
    .groupby('sexo_desc')['ingreso']
    .agg(
        Media_Promedio = 'mean',
        Mediana_Tipica = 'median'
    )
    .round(2)
)

print("\n--- REPORTE EJECUTIVO DE INGRESOS (BIG DATA) ---")
print(reporte_tendencia)
```

### El Procedimiento Senior: Motor de Auditor√≠a Estad√≠stica

En un ERP, no puedes tener a un gerente mirando 50 histogramas para adivinar si los datos est√°n tirados a la derecha o a la izquierda. Necesitas un **Auditor Estad√≠stico** automatizado que procese todas las columnas num√©ricas y diagnostique su forma usando el **Coeficiente de Asimetr√≠a (Skewness)**.

Aqu√≠ tienes la refactorizaci√≥n resolviendo la deuda t√©cnica de repetici√≥n (DRY) y aplicando l√≥gica program√°tica tolerante a fallos.

```python

# ==============================================================================
# MOTOR DE AUDITOR√çA ESTAD√çSTICA (BACKEND)
# ==============================================================================

def auditor_estadistico(df: pd.DataFrame, columnas: list) -> pd.DataFrame:
    """
    Eval√∫a la tendencia central y la asimetr√≠a matem√°tica de m√∫ltiples variables.
    Ideal para perfilado de datos (Data Profiling) automatizado.
    """
    resultados = []

    for col in columnas:
        # 1. Extracci√≥n de M√©tricas (Dejamos que Pandas trabaje en C)
        media = df[col].mean()
        mediana = df[col].median()
        # Tomamos la moda principal para evitar el error de promediar variables multimodales
        moda_principal = df[col].mode().iloc[0]

        # 2. T√âCNICA SENIOR: Coeficiente de Asimetr√≠a (Skewness)
        # Calcula matem√°ticamente hacia d√≥nde se estira la "cola" de los datos
        sesgo = df[col].skew()

        # 3. L√≥gica de Negocio Autom√°tica
        # No usamos una igualdad estricta (== 0) porque en datos reales nunca es exacto.
        # Usamos un umbral de tolerancia estad√≠stica (+/- 0.5)
        if sesgo > 0.5:
            forma = "Asimetr√≠a Positiva (Derecha)"
        elif sesgo < -0.5:
            forma = "Asimetr√≠a Negativa (Izquierda)"
        else:
            forma = "Aproximadamente Sim√©trica"

        resultados.append({
            'Variable': col.replace('_', ' ').capitalize(), # Limpieza visual
            'Media': round(media, 2),
            'Mediana': round(mediana, 2),
            'Moda (Principal)': round(moda_principal, 2),
            'Skewness': round(sesgo, 3),
            'Diagn√≥stico': forma
        })

    # Retornamos un DataFrame estructurado y listo para el frontend
    return pd.DataFrame(resultados).set_index('Variable')

# ==============================================================================
# EJECUCI√ìN DEL REPORTE EJECUTIVO
# ==============================================================================

# Definimos las variables a auditar de una sola vez
variables_interes = ['ingreso', 'altura', 'a√±os_de_estudio']

reporte_auditoria = auditor_estadistico(datos, variables_interes)

print("--- REPORTE DE PERFILADO DE DATOS (DATA PROFILING) ---")
display(reporte_auditoria)
```

**¬øQu√© logramos con esta salida?**
En lugar de tres bloques de c√≥digo separados y comprobaciones manuales fr√°giles (`if moda < mediana < media`), el c√≥digo ahora devuelve una tabla limpia de 3 filas. El sistema detecta que el `Ingreso` tiene un `Skewness` fuertemente positivo (arrastrado por los millonarios), que la `Altura` est√° muy cerca de 0 (sim√©trica), y que los `A√±os de estudio` tienen un valor negativo (arrastrados por la mayor√≠a que termin√≥ la escuela, frente a unos pocos que no estudiaron).

#### üìù Apunte: Asimetr√≠a y el Coeficiente Skewness

La regla emp√≠rica de relacionar Moda, Mediana y Media ($Mo < Md < \bar{X}$) es excelente para la teor√≠a, pero se rompe f√°cilmente en la pr√°ctica si los datos son multimodales o discretos. Para resolver esto, los Cient√≠ficos de Datos usan el **Coeficiente de Asimetr√≠a de Fisher-Pearson (Skewness)**.

La f√≥rmula matem√°tica mide el tercer momento estandarizado de la distribuci√≥n:

$$Skewness = \frac{n}{(n-1)(n-2)} \sum \left( \frac{x_i - \bar{x}}{s} \right)^3$$

**Interpretaci√≥n en L√≥gica de Negocio (Backend):**
Al integrar `.skew()` en tu sistema, puedes programar "alertas autom√°ticas" en tu ERP. Por ejemplo:

* Si `skew() > 0.5` (Asimetr√≠a Positiva): El sistema bloquea el uso de la Media en el dashboard de ventas e imprime la Mediana por defecto, protegiendo al gerente de tomar malas decisiones por culpa de un cliente at√≠pico.
* Si `skew() < -0.5` (Asimetr√≠a Negativa): El sistema sabe que la mayor√≠a de los usuarios est√°n acumulados en la parte alta del rango (por ejemplo, evaluaciones de desempe√±o donde casi todos sacan 10).
* Si `skew()` est√° entre $-0.5$ y $0.5$: Los datos son sim√©tricos (normales) y es seguro usar el Promedio tradicional.

### üìù Etapa 1: Apunte Te√≥rico - Medidas de Localizaci√≥n en Negocios

A diferencia de las medidas de tendencia central (que buscan el "medio"), las **Medidas de Localizaci√≥n (Cuantiles)** buscan dividir tu base de datos en porciones exactas para entender los extremos y las transiciones.

En un CRM , esto es vital para el **Scoring de Clientes**:

* **Cuartiles (Dividen en 4):** * **Q1 (25%):** El l√≠mite de tus clientes "Base" o sensibles al precio. Si el ingreso cae aqu√≠, el riesgo de *Churn* (fuga por precio) es alto.
* **Q2 (50%):** Es exactamente la Mediana. Tu cliente "T√≠pico".
* **Q3 (75%):** Tu segmento "Premium". Los que superan este l√≠mite son tu mercado objetivo para ventas cruzadas (*Cross-selling*).


* **Deciles (Dividen en 10):** Permiten aislar al **Decil 9 (Top 10%)**, que estad√≠sticamente en ventas B2B (empresas) suele generar el 80% de tus ingresos reales.
* **Percentiles (Dividen en 100):** Usados para micro-segmentaci√≥n o para detectar anomal√≠as (ej: "Aislar al percentil 99 porque son errores de facturaci√≥n o fraudes").

---

### üíª Etapa 1: C√≥digo Senior (Motor de Cuantiles)

El problema del c√≥digo del curso (`[i/10 for i in range(1,10)]`) es que devuelve una serie de n√∫meros sin contexto. Un sistema no puede leer eso y un gerente no sabe qu√© significa "0.3".

Vamos a aplicar **Clean Code** creando una funci√≥n robusta que traduzca la matem√°tica a lenguaje de negocio, validando los datos de entrada y devolviendo un DataFrame profesional.

---

```python
# ==============================================================================
# MOTOR DE SEGMENTACI√ìN B√ÅSICO (BACKEND)
# ==============================================================================

def generar_reporte_cuantiles(serie: pd.Series, tipo: str = 'cuartil') -> pd.DataFrame:
    """
    Calcula los l√≠mites de segmentaci√≥n de una variable num√©rica.
    Previene deuda t√©cnica devolviendo un DataFrame estructurado y legible.
    """
    # 1. Validaci√≥n de entrada (Clean Code)
    if serie.empty:
        raise ValueError("Error: La serie de datos est√° vac√≠a.")

    # 2. Diccionario de Arquitectura (Evita anidar m√∫ltiples if/else)
    configuracion = {
        'cuartil': {
            'cortes': [0.25, 0.50, 0.75],
            'etiquetas': ['Q1 (Base 25%)', 'Q2 (Mediana 50%)', 'Q3 (Premium 75%)']
        },
        'decil': {
            'cortes': [i/10 for i in range(1, 10)],
            'etiquetas': [f'D{i} (Acumulado {i}0%)' for i in range(1, 10)]
        }
    }

    if tipo not in configuracion:
        raise ValueError("Tipo no soportado. Elija 'cuartil' o 'decil'.")

    # 3. Extracci√≥n de Par√°metros
    cortes = configuracion[tipo]['cortes']
    etiquetas = configuracion[tipo]['etiquetas']

    # 4. C√°lculo Vectorizado (Delega a Pandas)
    valores_limite = serie.quantile(cortes)

    # 5. Capa de Presentaci√≥n (Ensamblaje del DataFrame)
    df_segmentos = pd.DataFrame({
        'Corte Estrat√©gico': etiquetas,
        'L√≠mite de Ingreso ($)': valores_limite.values
    })

    # Seteamos el √≠ndice para que sea visualmente profesional y f√°cil de consultar
    return df_segmentos.set_index('Corte Estrat√©gico').round(2)

# --- PRUEBA DEL MOTOR ---
# En lugar de imprimir listas crudas, generamos tablas de negocio
reporte_cuartiles = generar_reporte_cuantiles(datos['ingreso'], tipo='cuartil')
reporte_deciles = generar_reporte_cuantiles(datos['ingreso'], tipo='decil')

print("--- REPORTE ESTRAT√âGICO: CUARTILES ---")
display(reporte_cuartiles)

print("\n--- REPORTE ESTRAT√âGICO: DECILES ---")
display(reporte_deciles)
```

El Box-Plot es el "detector de anomal√≠as" definitivo. Nos permite ver de un vistazo la dispersi√≥n de los datos y detectar si hay sesgos hacia alg√∫n extremo.

### üìù Apunte Te√≥rico: El "Bug" de las Desviaciones Absolutas

En la anal√≠tica de datos, existe una confusi√≥n muy com√∫n porque hay dos m√©tricas distintas que en ingl√©s comparten las mismas siglas (MAD). El curso calcula una de forma manual, pero luego usa una librer√≠a (SciPy) que calcula la otra, trat√°ndolas como si fueran lo mismo.

Para tu proyecto "Gesti√≥n PyME Inteligente", la diferencia entre ambas es la diferencia entre un reporte sesgado y uno real:

#### 1. Desviaci√≥n Media Absoluta (Mean Absolute Deviation)

Mide en promedio cu√°nto se aleja cada dato de la **Media** (el Promedio).

* **F√≥rmula:** $DM = \frac{1}{n}\sum_{i=1}^{n}|X_i - \bar{X}|$
* **L√≥gica de Negocio:** Es intuitiva ("en promedio, mis ventas var√≠an en \$ 2.000"), pero hereda el gran defecto de la Media: si tienes un solo mes con ventas at√≠picas enormes, la desviaci√≥n se infla artificialmente. Debido a esta fragilidad, los desarrolladores de Pandas **eliminaron** la funci√≥n `.mad()` de la librer√≠a.

#### 2. Desviaci√≥n Mediana Absoluta (Median Absolute Deviation)

Mide la distancia t√≠pica de los datos respecto a la **Mediana**. Es decir, ordena las distancias y toma la del medio.

* **F√≥rmula:** $MAD = Mediana(|X_i - \tilde{X}|)$
* **L√≥gica de Negocio:** Es **Estad√≠stica Robusta**. Ignora por completo los meses con ventas rid√≠culamente altas o bajas y te dice cu√°l es la variabilidad real de tu "Core Business". Esta es exactamente la m√©trica que calcula `scipy.stats.median_abs_deviation`.

### üíª Etapa Senior: Auditor de Riesgo Operativo (Clean Code)

En producci√≥n no creamos columnas temporales (`['desviacion']`, `['desviacion_absoluta']`) en el DataFrame original porque eso ensucia la base de datos y consume memoria innecesaria. En su lugar, usamos **Vectorizaci√≥n** pura con NumPy y Pandas para hacer el c√°lculo al vuelo.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# ==============================================================================
# MOTOR DE DISPERSI√ìN (AUDITOR DE RIESGO OPERATIVO)
# ==============================================================================

def auditar_riesgo_operativo(serie: pd.Series, nombre_entidad: str) -> dict:
    """
    Calcula las m√©tricas de dispersi√≥n absoluta para evaluar la estabilidad.
    Distingue claramente entre la desviaci√≥n respecto a la Media y a la Mediana.
    """
    if serie.empty:
        raise ValueError("La serie no contiene datos.")

    # 1. Desviaci√≥n Media Absoluta (C√°lculo manual vectorizado, ya que Pandas lo deprec√≥)
    # serie.mean() calcula el promedio. np.abs() convierte los negativos en positivos.
    media_absoluta = np.abs(serie - serie.mean()).mean()

    # 2. Desviaci√≥n Mediana Absoluta (Estad√≠stica Robusta usando SciPy)
    # scale='normal' ajusta el valor para que sea comparable con la desviaci√≥n est√°ndar
    mediana_absoluta = stats.median_abs_deviation(serie, scale='normal')

    return {
        'Entidad': nombre_entidad,
        'Desviaci√≥n (vs Promedio)': round(media_absoluta, 2),
        'Desviaci√≥n (vs Mediana)': round(mediana_absoluta, 2)
    }

# --- EJECUCI√ìN DEL REPORTE EJECUTIVO ---
# Comparamos la "estabilidad" de Mar√≠a vs Pablo
reporte_maria = auditar_riesgo_operativo(df['Mar√≠a'], 'Mar√≠a (Riesgo Alto)')
reporte_pablo = auditar_riesgo_operativo(df['Pablo'], 'Pablo (Riesgo Bajo)')

# Ensamblamos los resultados en un DataFrame limpio
df_riesgo = pd.DataFrame([reporte_maria, reporte_pablo]).set_index('Entidad')

print("--- REPORTE DE VOLATILIDAD / RIESGO OPERATIVO ---")
display(df_riesgo)
```

### üìù Apunte Te√≥rico: Varianza y Desviaci√≥n Est√°ndar

En el mundo corporativo, estas dos m√©tricas son los indicadores oficiales de **Riesgo y Volatilidad**.

#### 1. Varianza ($S^2$ o $\sigma^2$)

Mide el promedio de las desviaciones al cuadrado respecto a la media.

* **Poblaci√≥n:** $\sigma^2 = \frac{1}{N}\sum_{i=1}^{N}(X_i-\mu)^2$
* **Muestra:** $S^2 = \frac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$ *(Se divide por $n-1$, Correcci√≥n de Bessel, para no subestimar el error)*.
* **Problema de Negocio:** Sus unidades est√°n al cuadrado. Si analizas el ingreso de tu PyME, la varianza te dar√° el resultado en **"Pesos al Cuadrado"**, lo cual es imposible de interpretar en una reuni√≥n gerencial.

#### 2. Desviaci√≥n Est√°ndar ($S$ o $\sigma$)

Es simplemente la ra√≠z cuadrada de la Varianza.

* **Poblaci√≥n:** $\sigma = \sqrt{\sigma^2}$
* **Muestra:** $S = \sqrt{S^2}$
* **Ventaja de Negocio:** Te devuelve el indicador a su unidad original. Si la media de ventas es $\$10.000$ y la desviaci√≥n est√°ndar es $\$2.000$, sabes exactamente en qu√© rango monetario se mueve tu empresa.

### üíª Etapa Senior: Motor de Volatilidad (Clean Code)

A continuaci√≥n, crearemos un bloque de c√≥digo profesional que audita las notas sin ensuciar el DataFrame original, usando vectorizaci√≥n de NumPy y los par√°metros expl√≠citos de Pandas.



```python


import pandas as pd
import numpy as np

# ==============================================================================
# MOTOR DE VOLATILIDAD (LABORATORIO MATEM√ÅTICO VS PRODUCCI√ìN)
# ==============================================================================

def auditar_volatilidad(serie: pd.Series, nombre: str, es_muestra: bool = True) -> dict:
    """
    Calcula Varianza y Desviaci√≥n Est√°ndar de forma manual (vectorizada) y
    compara con el motor de C de Pandas para validar la matem√°tica.
    """
    if serie.empty:
        raise ValueError("La serie est√° vac√≠a.")

    n = len(serie)
    media = serie.mean()

    # 1. EL CAMINO MANUAL VECTORIZADO (Sin crear columnas en el DataFrame)
    # Restamos la media, elevamos al cuadrado y sumamos de un solo golpe.
    suma_desviaciones_cuadradas = np.sum((serie - media)**2)

    # L√≥gica de Grados de Libertad (n vs n-1)
    denominador = (n - 1) if es_muestra else n

    varianza_manual = suma_desviaciones_cuadradas / denominador
    std_manual = np.sqrt(varianza_manual)

    # 2. EL CAMINO DE PRODUCCI√ìN (Pandas robusto)
    # T√âCNICA SENIOR: ddof (Delta Degrees of Freedom).
    # ddof=1 es Muestra (n-1). ddof=0 es Poblaci√≥n (N).
    grados_libertad = 1 if es_muestra else 0

    varianza_pandas = serie.var(ddof=grados_libertad)
    std_pandas = serie.std(ddof=grados_libertad)

    return {
        'Estudiante/Entidad': nombre,
        'Media (Promedio)': round(media, 2),
        'Varianza (Unidades cuadradas)': round(varianza_pandas, 2),
        'Desv. Est√°ndar (Unidades reales)': round(std_pandas, 2)
    }

# --- EJECUCI√ìN DEL REPORTE EJECUTIVO ---
# Evaluamos a todos los estudiantes de una sola vez
resultados = [
    auditar_volatilidad(df['Mar√≠a'], 'Mar√≠a'),
    auditar_volatilidad(df['Pablo'], 'Pablo'),
    auditar_volatilidad(df['Pedro'], 'Pedro')
]

# Ensamblamos el reporte final
reporte_dispersion = pd.DataFrame(resultados).set_index('Estudiante/Entidad')

print("--- REPORTE DE VOLATILIDAD Y DESEMPE√ëO ---")
display(reporte_dispersion)
```
