### üåê Ingesta de Datos Web: Formato JSON

**Concepto de Negocio:**
JSON es el formato est√°ndar para el intercambio de datos en la web (APIs). A diferencia de las bases de datos relacionales, JSON no es tabular (filas/columnas), sino jer√°rquico (diccionarios anidados).

**El Problema del Anidamiento:**
La funci√≥n `pd.read_json()` funciona perfectamente para estructuras planas. Sin embargo, si el archivo contiene "diccionarios dentro de diccionarios" (datos anidados), Pandas comprimir√° el sub-diccionario en una sola celda de texto, imposibilitando el an√°lisis matem√°tico o estad√≠stico. A este proceso de aplanar la jerarqu√≠a se le conoce como **Normalizaci√≥n**.

```python
import pandas as pd

# 1. Definici√≥n de Rutas (Tu Google Drive)
RUTA_JSON_PLANO = '/content/drive/MyDrive/Pandas/pacientes.json'
RUTA_JSON_ANIDADO = '/content/drive/MyDrive/Pandas/pacientes_2.json'

# -------------------------------------------------------------------
# 2. El Escenario Feliz (JSON Plano)
# -------------------------------------------------------------------
df_pacientes_plano = pd.read_json(RUTA_JSON_PLANO)

print("‚úÖ Ingesta Exitosa: JSON Plano")
print("--- Vista de Datos Limpios ---")
display(df_pacientes_plano.head(3))
print("-" * 50)


# -------------------------------------------------------------------
# 3. El Escenario del Mundo Real (JSON Anidado)
# -------------------------------------------------------------------
df_pacientes_anidado = pd.read_json(RUTA_JSON_ANIDADO)

print("\n‚ö†Ô∏è Advertencia de Ingesta: JSON Anidado detectado")
print("--- Vista de Datos Compresos (Requieren Normalizaci√≥n) ---")
display(df_pacientes_anidado.head(3))
```

### üîë Asignaci√≥n de √çndice L√≥gico (Primary Key)

**Concepto de Negocio:**
Cuando un archivo origen ya trae un identificador √∫nico (como un DNI, un CUIT o un `ID_paciente`), dejar que Pandas cree un √≠ndice num√©rico adicional es redundante y consume memoria.

**Estrategia de Soluci√≥n:**
Debemos promover esa columna identificadora para que se convierta en el √≠ndice oficial del DataFrame.
* En archivos planos (CSV/Excel), usamos el par√°metro `index_col`.
* En archivos jer√°rquicos (JSON), encadenamos el m√©todo `.set_index('nombre_columna')` inmediatamente despu√©s de la lectura.

```python
import pandas as pd

RUTA_JSON_PLANO = '/content/drive/MyDrive/Pandas/pacientes.json'

# Leemos el JSON y autom√°ticamente promovemos 'ID_paciente' como el √≠ndice oficial.
df_pacientes_plano = pd.read_json(RUTA_JSON_PLANO).set_index('ID_paciente')

print("‚úÖ Ingesta Exitosa: √çndice l√≥gico asignado")
display(df_pacientes_plano.head(3))
```

##**Normalizando archivos JSON**

### ‚úÇÔ∏è Normalizaci√≥n por Extracci√≥n Directa (M√©todo R√°pido)

**Concepto de Negocio:**
Cuando un archivo JSON tiene una estructura anidada dentro de una tabla que ya cargamos en memoria, podemos extraer exclusivamente esa columna problem√°tica y "aplanarla".

**Mec√°nica del C√≥digo:**
El instructor utiliza `pd.json_normalize(datos_pacientes_2['Pacientes'])`.
1. `datos_pacientes_2` es nuestro DataFrame original con el formato roto.
2. `['Pacientes']` a√≠sla espec√≠ficamente la columna que contiene los diccionarios anidados (las listas de pacientes).
3. `json_normalize()` toma esa Serie aislada y la expande convirtiendo cada llave del diccionario interno en una nueva columna independiente (Raza, IMC, Fumador, etc.).

**‚ö†Ô∏è Riesgo Arquitect√≥nico (Trade-off):**
Este m√©todo es r√°pido y directo, pero aplica una "visi√≥n de t√∫nel". Al aislar √∫nicamente la columna `'Pacientes'`, el resultado final descarta cualquier metadato global que existiera en las otras columnas del archivo original (como el A√±o de la investigaci√≥n o el Nombre del proyecto). Es √∫til para an√°lisis r√°pidos y aislados, pero no se recomienda si necesitamos mantener la trazabilidad completa (Data Lineage) en bases de datos relacionales.

```python
# para ver la estructura del json podemos verlo en https://jsoncrack.com/
df_normalizado = pd.json_normalize(datos_pacientes_2['Pacientes'])
```

### üõ†Ô∏è Normalizaci√≥n Avanzada de JSON (Preservando Integridad)

**El Riesgo de Negocio:**
El archivo `pacientes_2.json` contiene metadatos globales en su ra√≠z (`Pesquisa` y `A√±o`) y una lista anidada en la llave `pacientes`. Si normalizamos atacando √∫nicamente la columna 'pacientes' de un DataFrame pre-cargado, perderemos el contexto hist√≥rico del archivo.

**Estrategia de Soluci√≥n (El Camino Senior):**
1. **Lectura Nativa:** En lugar de forzar a Pandas a leer el archivo de golpe, usamos la librer√≠a est√°ndar `json` de Python para abrir el archivo de forma segura y cargarlo en la memoria como un diccionario puro.
2. **Despliegue Quir√∫rgico (`json_normalize`):** Le pasamos ese diccionario a Pandas.
   * `record_path`: Le indicamos qu√© lista interna queremos "desempaquetar" para crear las filas (los pacientes).
   * `meta`: Le indicamos qu√© datos de la "cabecera" queremos copiar y pegar en cada una de esas nuevas filas para no perder la trazabilidad.

```python
# Bloque de Diagn√≥stico: Inspeccionando las llaves reales
print("Las verdaderas llaves (keys) en la ra√≠z del archivo son:")
print(diccionario_datos.keys())
```

```python
import pandas as pd
import json

RUTA_JSON_ANIDADO = '/content/drive/MyDrive/Pandas/pacientes_2.json'

# 1. Cargamos el JSON crudo en memoria
with open(RUTA_JSON_ANIDADO, 'r', encoding='utf-8') as archivo_json:
    diccionario_datos = json.load(archivo_json)

# 2. Normalizaci√≥n con Integridad de Metadatos
df_pacientes_normalizado = pd.json_normalize(
    diccionario_datos,
    record_path=['Pacientes'],
    meta=['Investigaci√≥n', 'A√±o']
)

# Revision nombres de columnas
print("üîç Las verdaderas columnas generadas son:")
print(df_pacientes_normalizado.columns.tolist())

# 3. Asignamos el √≠ndice l√≥gico usando el nombre REAL de la columna
df_pacientes_normalizado = df_pacientes_normalizado.set_index('ID')

print("‚úÖ Normalizaci√≥n Exitosa: Datos desempaquetados, metadatos conservados e √≠ndice corregido.")
display(df_pacientes_normalizado.head(3))
```

### üìä Auditor√≠a del Resultado

Al aplicar `record_path` y `meta`, hemos logrado "aplanar" la estructura tridimensional del JSON en una tabla bidimensional perfecta para el an√°lisis relacional.
* Los pacientes ya no est√°n atrapados en listas.
* La columna problem√°tica original se ha resuelto.
* **Gran Diferencia:** Si nos desplazamos hasta las √∫ltimas columnas a la derecha del DataFrame, veremos que Pandas cre√≥ autom√°ticamente las columnas `Investigaci√≥n` y `A√±o` y les asign√≥ el valor correspondiente a cada paciente. La trazabilidad de los datos est√° asegurada al 100%.

### üì§ Exportaci√≥n a JSON y la regla de la Orientaci√≥n

**Concepto de Negocio:**
Al igual que con CSV y Excel, la consistencia de la API de Pandas nos permite exportar datos f√°cilmente usando `to_json()`. Sin embargo, la forma en que estructuramos ese archivo de salida es cr√≠tica si va a ser consumido por otras aplicaciones (APIs, Front-End, bases de datos NoSQL).

**Buenas Pr√°cticas (El par√°metro `orient`):**
Por defecto, Pandas exporta los JSON agrupados por columnas (`orient='columns'`). En la industria del software, el est√°ndar es agrupar por entidades o filas (`orient='records'`). Esto genera una lista limpia de diccionarios, donde cada diccionario representa un √∫nico registro (ej. un paciente o una factura), facilitando la integraci√≥n con cualquier otro sistema.

```python
import pandas as pd

# 1. Definici√≥n de Rutas
RUTA_EXPORTACION_JSON = '/content/drive/MyDrive/Pandas/historico_pacientes_limpio.json'

# 2. Exportaci√≥n Senior (Recuperando el ID y Orientando por Registros)
# Encadenamos .reset_index() para que el 'ID' vuelva a ser una columna exportable
df_pacientes_normalizado.reset_index().to_json(
    RUTA_EXPORTACION_JSON,
    orient='records',
    force_ascii=False,
    indent=4
)

print(f"‚úÖ Archivo JSON exportado con √©xito preservando el ID en: {RUTA_EXPORTACION_JSON}")

# 3. Auditor√≠a de Verificaci√≥n
df_verificacion_json = pd.read_json(RUTA_EXPORTACION_JSON)

# Para la visualizaci√≥n, volvemos a poner el ID como √≠ndice l√≥gico para leerlo bien
df_verificacion_json = df_verificacion_json.set_index('ID')

print("\n--- Vista previa del archivo exportado (¬°ID Recuperado!) ---")
display(df_verificacion_json.head(3))
```

### üåê Conexi√≥n en Vivo: Consumo de APIs REST (JSON)

**Concepto de Negocio:**
Las APIs permiten que dos sistemas se comuniquen en tiempo real. En este caso, nuestro sistema en Python se conectar√° a la base de datos externa de *Fruityvice* para alimentar nuestra aplicaci√≥n con datos nutricionales actualizados.

**Mejores Pr√°cticas (Arquitectura Senior):**
1. **Petici√≥n HTTP:** Utilizamos la librer√≠a `requests` ([Documentacion](https://pypi.org/project/requests/)) para hacer una llamada tipo `GET` a la URL.
2. **Extracci√≥n Nativa:** Usamos el m√©todo `.json()` de la respuesta para evitar importar la librer√≠a `json` innecesariamente.
3. **Normalizaci√≥n (Correcci√≥n de QA):** En lugar de usar `pd.DataFrame()`, aplicamos nuestra herramienta `pd.json_normalize()` para aplanar el diccionario anidado de `nutritions` y convertir los carbohidratos, prote√≠nas y calor√≠as en columnas reales y analizables.

```python
import pandas as pd
import requests

# 1. Definici√≥n del Endpoint (La URL donde atiende el "mesero")
URL_API_FRUTAS = 'https://fruityvice.com/api/fruit/all'

# 2. Extracci√≥n (Petici√≥n HTTP GET)
print("Conectando con la API de Fruityvice...")
respuesta = requests.get(URL_API_FRUTAS)

# Control de Calidad: Verificamos que el servidor nos dio permiso (C√≥digo 200 = OK)
if respuesta.status_code == 200:

    # 3. Transformaci√≥n (Parseo y Normalizaci√≥n)
    # Extraemos el JSON directamente del objeto de respuesta
    datos_json_crudos = respuesta.json()

    # Normalizamos para desempaquetar las calor√≠as y vitaminas
    df_frutas_normalizado = pd.json_normalize(datos_json_crudos)

    # Asignamos el ID l√≥gico
    df_frutas_normalizado = df_frutas_normalizado.set_index('id')

    print("‚úÖ Ingesta exitosa. Datos nutricionales normalizados.")
    display(df_frutas_normalizado.head())

else:
    # Si recibimos un 403 (Forbidden) u otro error, el sistema nos avisa
    print(f"‚ùå Error al conectar a la API. C√≥digo de estado HTTP: {respuesta.status_code}")
```

### üïµÔ∏è‚Äç‚ôÇÔ∏è An√°lisis de C√≥digo (Legacy): El M√©todo del Curso

**Concepto T√©cnico:**
El material de apoyo presenta una sintaxis que, aunque funcional, en la industria actual se considera "deuda t√©cnica" (c√≥digo sub√≥ptimo) por dos razones arquitect√≥nicas:

1. **Parseo Manual Redundante:** El curso importa la librer√≠a `json` y utiliza `json.loads(respuesta.text)` para traducir el texto plano de la API. Esto era est√°ndar hace a√±os, pero las versiones modernas de la librer√≠a `requests` incluyen el m√©todo nativo `.json()`, haciendo que la importaci√≥n de una segunda librer√≠a sea innecesaria.
2. **Falta de Normalizaci√≥n (El error de Negocio):** Al pasar el objeto directamente a `pd.DataFrame(resultado)`, Pandas trata los datos de forma bidimensional estricta. Como resultado, la llave jer√°rquica `nutritions` queda atrapada como un diccionario dentro de una sola celda (ej. `{'calories': 52, 'fat': 0.4}`). Esto hace imposible que nuestro sistema calcule promedios o filtre por calor√≠as sin realizar transformaciones complejas posteriores.
