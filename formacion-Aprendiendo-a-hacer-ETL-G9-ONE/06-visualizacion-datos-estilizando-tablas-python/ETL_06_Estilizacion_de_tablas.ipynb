### 🗄️ Ingesta y Preparación de Datos (ETL): Enfoque Senior

**Contexto del Negocio:**
Consumimos un dataset transaccional de ventas (Tickets). El objetivo es preparar esta sábana de datos para la construcción de reportes tabulares de alto nivel gerencial.

**Arquitectura del Código (Method Chaining):**
En lugar de mutar el DataFrame paso a paso a lo largo de múltiples celdas (lo que genera riesgos si se ejecutan en desorden), aplicamos el patrón de "Encadenamiento de Métodos". Esto consolida la extracción, el filtrado de columnas (Feature Selection) y el parseo de tipos de datos en una sola transacción atómica y eficiente.

````python
import pandas as pd

# ==============================================================================
# PIPELINE DE INGESTA Y PREPARACIÓN (Method Chaining)
# ==============================================================================
print("🚀 Ejecutando Pipeline de Extracción y Transformación...")

url_datos = 'https://gist.githubusercontent.com/ahcamachod/d2efadda3f3833d65da750c64705ac42/raw/80aef6f3bad210d34b67ed85565f039a30e2d9e0/ventas.csv'

# Definimos las columnas estrictamente necesarias para el modelo de negocio
columnas_negocio = [
    'fecha_pedido', 'modo_envio', 'nombre_cliente', 'segmento_cliente',
    'ciudad', 'provincia', 'region', 'departamento', 'tipo_producto',
    'ventas', 'cantidad', 'ganancia'
]

# Ejecutamos la lectura, el filtrado y la conversión de fechas en un solo bloque
df_ventas = (
    pd.read_csv(url_datos, sep='#')
    .loc[:, columnas_negocio]                                      # 1. Filtramos las columnas
    .assign(fecha_pedido=lambda x: pd.to_datetime(x['fecha_pedido'])) # 2. Parseo seguro de Fechas
)

# ==============================================================================
# AUDITORÍA DE DATOS
# ==============================================================================
print("\n✅ DataFrame 'df_ventas' instanciado con éxito.")
print("-" * 50)
print(f"Dimensión de la matriz : {df_ventas.shape[0]} Filas x {df_ventas.shape[1]} Columnas")
print("-" * 50)

# Mostramos la cabecera para confirmar la estructura
display(df_ventas.head(3))
````

# Misión 1: Equipo de Ventas

Necesitamos construir una visualización para indicar los **10 principales clientes por total de ventas**.

Con ello, el equipo de ventas puede pensar en estrategias para mantener a estos clientes comprando en nuestra tienda.

````python
# Crearemos un nuevo dataframe con los principales clientes
# Nuestro grooupby por el nombre_cliente , haciendo la sum de todas las ventas ,
# y aplicando nlargest conseguimos los 10 mayores. a la vez hacemos una copia
# (defensivo). No olvidar de resetear el indice
clientes = ventas.groupby(['nombre_cliente'])['ventas'].sum().nlargest(10).copy()
clientes = clientes.reset_index()
clientes
````

````python
# El Renombrado de las columnas usando diccionario se hace de esta forma, un
# diccionario con llave el nombre viejo , y valor el nombre nuevo, luego
# utilizamos .rename(columns=diccionario)
# dic = {'nombre_cliente':'Clientes', 'ventas':'Ventas'}
# clientes = clientes.rename(columns=dic)

#Mejoraremos con la forma mas sencilla los nombres de las columnas
clientes.columns = ['Clientes','Ventas']
#Crearemos una columna adicional llamada ranking , que empieza en 1
clientes['Ranking'] = clientes.index + 1
#Luego cambiamos nuestra columna ranking para que sea nuestro index
clientes.set_index('Ranking', inplace = True)
clientes
````

### 🏆 Misión 1: Pipeline de Agregación (Top 10 Clientes)

**Contexto del Negocio:**
El equipo de ventas requiere identificar a los 10 clientes con mayor volumen de facturación histórica para integrarlos a un programa de fidelización (VIP).

**Arquitectura del Código (Clean Code):**
* **Split-Apply-Combine:** Utilizamos `.groupby()` para segmentar las ventas por cliente y `.sum()` para agregar el ingreso total.
* **Selección Eficiente:** `.nlargest()` es computacionalmente superior a ordenar todo el dataset (`.sort_values()`) y luego cortar los primeros 10 con `.head()`.
* **Renombrado Seguro:** El método `.rename(columns={...})` mediante diccionario garantiza que los nombres cambien correctamente sin depender del orden posicional de las columnas.
* **Índice Semántico:** Reemplazamos el índice autogenerado (0, 1, 2...) por un rango del 1 al 10, dándole el nombre semántico de `Ranking` para prepararlo para la presentación visual.

````python
# ==============================================================================
# PIPELINE DE AGREGACIÓN: TOP 10 CLIENTES (Enfoque Senior)
# ==============================================================================
print("🥇 Generando Ranking de Cuentas Clave (Key Accounts)...")

# Encadenamiento de métodos para transformación segura
df_top_clientes = (
    df_ventas
    .groupby('nombre_cliente', as_index=False)['ventas'].sum()  # Agrupamos y sumamos
    .nlargest(10, columns='ventas')                             # Filtramos los 10 mayores
    .rename(columns={'nombre_cliente': 'Cliente', 'ventas': 'Ventas'}) # Renombramos
)

# Ajuste del Índice para que funcione como nuestro Ranking (1 al 10)
df_top_clientes.index = range(1, 11)
df_top_clientes.index.name = 'Ranking'

# Visualización del resultado crudo (sin estilizar aún)
display(df_top_clientes)
````

### 🎨 La Capa de Presentación: El Objeto `Styler` y Formateo Numérico

**Contexto Tecnológico:**
Al invocar `df.style`, Pandas retorna un objeto `Styler`. Este objeto expone una API para inyectar reglas de diseño (CSS) y transformaciones de texto que modifican **únicamente la visualización HTML** de la tabla, sin alterar los datos matemáticos subyacentes.

**Diseño Financiero (String Formatting):**
Para los reportes de ventas, utilizamos el micro-lenguaje de formateo de Python.
Desglose de la regla `"${:,.2f}"`:
* `$`: Símbolo literal de moneda que aparecerá al inicio.
* `:`: Indica el comienzo de la regla de formato.
* `,`: Instruye al motor a usar comas como separadores de miles.
* `.2f`: Limita la precisión a 2 decimales flotantes (*float*), estandarizando los centavos.

````python
# ==============================================================================
# ESTILIZACIÓN Y FORMATEO FINANCIERO (Capa de Presentación)
# ==============================================================================
print("💎 Renderizando Reporte Financiero para el Equipo de Ventas...")

# Definimos las reglas de formato en un diccionario para mantener el código limpio
formato_columnas = {
    'Ventas': '${:,.2f}' # Agregamos el símbolo $ para contexto de negocios
}

# Aplicamos el estilo directamente encadenando el método .format()
reporte_ventas_estilizado = df_top_clientes.style.format(formato_columnas)

# Desplegamos el objeto Styler final
display(reporte_ventas_estilizado)
````

### 📤 Exportación de Reportes Estilizados (Objetos `Styler`)

**Contexto Tecnológico:**
Una vez que inyectamos CSS y formatos financieros en nuestra tabla con `.style`, esta deja de ser un DataFrame puro y se convierte en un objeto `Styler`. Pandas expone métodos nativos para renderizar este objeto directamente en los formatos ofimáticos y web más utilizados por los tomadores de decisiones.

**Métodos de Exportación Soportados:**
1. **`to_html()`:** Genera el código HTML puro con el CSS embebido. Ideal para incrustar el reporte en Dashboards web, correos electrónicos corporativos o intranets.
2. **`to_excel()`:** Renderiza la tabla en un archivo `.xlsx`. Mantiene los colores, formatos numéricos y estilos visuales, permitiendo que el área de finanzas continúe su análisis en Microsoft Excel. *(Requiere el motor `openpyxl`)*.
3. **`to_latex()`:** Transforma la tabla al lenguaje de marcado científico LaTeX. Imprescindible para publicaciones académicas y documentación técnica rigurosa.

> ⚠️ **Nota de Arquitectura:** Si necesitas exportar a un formato plano sin formato visual (como un `.csv`), no puedes usar el objeto `Styler`. Debes aplicar el método de exportación (`to_csv()`) directamente sobre el DataFrame original.

````python
import pandas as pd

# ==============================================================================
# DEMOSTRACIÓN: Exportación Multi-Formato de Objetos Styler
# ==============================================================================
print("🖨️ Generando archivos de reporte (HTML, Excel, LaTeX)...")

# 1. Dataset de prueba (Películas)
datos_peliculas = {
    'Género': ['Drama', 'Comedia', 'Acción', 'Ciencia Ficción', 'Terror'],
    'Nota Media': [4.5, 4.3, 4.2, 4.1, 3.8],
    'Cantidad de Evaluaciones': [1500, 1200, 1000, 800, 600]
}

# 2. Pipeline de Transformación (Method Chaining)
df_peliculas = (
    pd.DataFrame(datos_peliculas)
    # Calculamos el ranking denso basado en evaluaciones y lo forzamos a entero
    .assign(Ranking=lambda x: x['Cantidad de Evaluaciones'].rank(method='dense', ascending=False).astype(int))
    # Asignamos el ranking como índice para la presentación
    .set_index('Ranking')
)

# 3. Instanciamos el objeto Styler (Capa de Presentación)
reporte_estilizado = df_peliculas.style

# ==============================================================================
# 4. EXPORTACIÓN A DISCO
# ==============================================================================
# Web / Email
reporte_estilizado.to_html('reporte_peliculas.html')

# Microsoft Excel
reporte_estilizado.to_excel('reporte_peliculas.xlsx', engine='openpyxl')

# LaTeX (Documentación Académica/Técnica)
with open('reporte_peliculas.tex', 'w') as f:
    f.write(reporte_estilizado.to_latex())

print("✅ Exportación exitosa. Revisa el explorador de archivos de Colab.")
display(reporte_estilizado)
````

### 🎨 Teoría UI/UX: Data Storytelling con Colores

El color en un reporte financiero no cumple una función decorativa, sino **cognitiva**. Aplicar colores de forma aleatoria o excesiva distrae y confunde al tomador de decisiones. Para construir tablas efectivas, aplicamos las siguientes reglas de Arquitectura Visual:

1. **Semántica del Color:** Los colores deben asociarse a una narrativa universal.
   * **Rojo/Naranja (Colores Cálidos):** Representan alerta, peligro o pérdida financiera. Atraen inmediatamente el ojo humano.
   * **Verde/Azul (Colores Fríos):** Representan éxito, rentabilidad positiva o estabilidad.
2. **Reducción de Carga Cognitiva:** Evitar el uso de tablas estilo "arcoíris". Se debe usar un fondo neutro (blanco/gris) y aplicar color *únicamente* a los valores atípicos (máximos y mínimos) para guiar la vista directamente hacia donde se requiere acción gerencial.
3. **Ergonomía Visual (A11y):** No se deben utilizar colores puros o saturados (ej. rojo sangre o verde flúor) porque cansan la vista al leer números en pantalla. Es un estándar de la industria utilizar **tonos pastel (Soft Hex Codes)** para los fondos de las celdas, garantizando el contraste con la tipografía negra.

# ==============================================================================
# PIPELINE DE RENTABILIDAD Y FORMATO CONDICIONAL (Misión 2)
# ==============================================================================
print("📊 Generando Reporte de Rentabilidad por Tipo de Producto...")

# 1. Transformación de Datos (Agrupación y Renombrado Semántico)
df_rentabilidad = (
    df_ventas
    .groupby('tipo_producto')[['ventas', 'ganancia']].sum()
    .rename_axis('Tipo de Producto')                                      # Limpiamos el nombre del índice
    .rename(columns={'ventas': 'Ingresos Brutos', 'ganancia': 'Utilidad'}) # Lenguaje de Negocios
)

# 2. Definición de Paleta Corporativa (Hex Codes suaves para no cansar la vista)
COLOR_ALERTA = '#f8d7da' # Rojo pastel (Pérdidas / Mínimos)
COLOR_EXITO = '#d4edda'  # Verde pastel (Líderes / Máximos)

# 3. Capa de Presentación (Styler Chaining)
reporte_comercial_estilizado = (
    df_rentabilidad.style
    .format('USD {:,.2f}')                    # Aplicamos el formato moneda a toda la tabla
    .highlight_max(color=COLOR_EXITO)         # Iluminamos a los líderes (El que más vende y el que más gana)
    .highlight_min(color=COLOR_ALERTA)        # Iluminamos a los rezagados (El que menos vende y el que da pérdidas)
)

# Renderizado final
display(reporte_comercial_estilizado)

````python
# ==============================================================================
# PIPELINE DE RENTABILIDAD Y FORMATO CONDICIONAL (Misión 2)
# ==============================================================================
print("📊 Generando Reporte de Rentabilidad por Tipo de Producto...")

# 1. Transformación de Datos (Agrupación y Renombrado Semántico)
df_rentabilidad = (
    df_ventas
    .groupby('tipo_producto')[['ventas', 'ganancia']].sum()
    .rename_axis('Tipo de Producto')                                      # Limpiamos el nombre del índice
    .rename(columns={'ventas': 'Ingresos Brutos', 'ganancia': 'Utilidad'}) # Lenguaje de Negocios
)

# 2. Definición de CSS Corporativo (Fondo Pastel + Texto Oscuro forzado)
# Usamos el código #212529 (un gris casi negro muy usado en diseño web) para
# la legibilidad
CSS_ALERTA = 'background-color: #f8d7da; color: #212529;'
CSS_EXITO  = 'background-color: #d4edda; color: #212529;'

# 3. Capa de Presentación (Styler Chaining)
reporte_comercial_estilizado = (
    df_rentabilidad.style
    .format('USD {:,.2f}')                    # Aplicamos el formato moneda a toda la tabla
    .highlight_max(props=CSS_EXITO)           # Usamos 'props' en lugar de 'color'
    .highlight_min(props=CSS_ALERTA)       # Iluminamos a los rezagados (El que menos vende y el que da pérdidas)
)

# Renderizado final
display(reporte_comercial_estilizado)
````

### 🖥️ UI/UX Analítico: Heatmaps y Especificidad CSS

**Contexto del Negocio:**
Para un equipo comercial, un mapa de calor (Gradient) permite ver la distribución general del volumen de negocio, mientras que las alertas duras (Min/Max) indican dónde tomar acción inmediata.

**Arquitectura del Objeto Styler (Cascada CSS):**
1. **Encabezados Corporativos:** Utilizamos selectores HTML (`th` para Table Header) inyectando propiedades de alineación, fuentes sin serifas (Arial) y capitalización.
2. **Orden de Ejecución (El secreto del éxito):**
   * Primero aplicamos `.background_gradient()` para teñir toda la columna matemáticamente.
   * Segundo, aplicamos `.highlight_max()` y `.highlight_min()`. Al ejecutarse después, el motor CSS de Pandas permite que estas reglas sobrescriban el gradiente en las celdas extremas, garantizando que las alertas visuales no se pierdan en el mapa de calor.

````python
# ==============================================================================
# PIPELINE DE FRONT-END TABULAR (Heatmap + Encabezados Corporativos)
# ==============================================================================
print("✨ Renderizando Interfaz Tabular (Mapa de Calor Puro)...")

# 1. Definición de Reglas CSS (Estándar Pythonico: Lista de Diccionarios y Tuplas)
estilos_corporativos = [
    {
        'selector': 'th', # Afecta solo a los encabezados de la tabla
        'props': [
            ('font-family', 'Arial, sans-serif'),
            ('font-weight', 'bold'),
            ('text-align', 'center'),
            ('text-transform', 'capitalize'),
            ('background-color', '#343a40'), # Fondo gris oscuro elegante
            ('color', '#ffffff'),            # Letra blanca para contraste
            ('padding', '10px 20px')         # Aire entre el texto y el borde
        ]
    },
    {
        'selector': 'td', # Afecta a las celdas de datos
        'props': [('padding', '8px 15px')]
    }
]

# 2. Encadenamiento Maestro del Objeto Styler
reporte_comercial_heatmap = (
    df_rentabilidad.style
    .format('USD {:,.2f}')                                    # Formato financiero
    .background_gradient(cmap='Blues')                       # Mapa de calor orgánico
    .set_table_styles(estilos_corporativos, overwrite=False)  # Inyecta el marco y los encabezados
)

# Renderizado final
display(reporte_comercial_heatmap)
````

### 🚚 Misión 3: Distribución Logística (Data Bars in-cell)

El equipo de logística requiere optimizar la asignación de la flota de camiones. Para ello, necesitan conocer el "Market Share" (Porcentaje de Participación) de cada región sobre el total nacional de pedidos.

**Arquitectura del Código (Pandas Avanzado & In-cell Charts):**
* **Cálculo Nativo (`normalize=True`):** Evitamos exportar datos a NumPy para hacer matemáticas manuales. El método `.value_counts()` de Pandas calcula las frecuencias relativas automáticamente.
* **Fusión de Series (`pd.concat`):** Unimos el conteo absoluto (Nº Pedidos) con el conteo relativo (%) en un solo DataFrame de forma limpia.
* **Mini-gráficos (`style.bar`):** Inyectamos barras de progreso dentro de la columna de porcentajes. Esto convierte la tabla estática en un panel de control interactivo visualmente, permitiendo comparar proporciones en fracciones de segundo.

````python
import pandas as pd

# ==============================================================================
# PIPELINE DE DISTRIBUCIÓN LOGÍSTICA (Market Share)
# ==============================================================================
print("🚚 Calculando Densidad de Pedidos por Región...")

# 1. Transformación de Datos (Cálculo Vectorizado y Nativo)
conteo_absoluto = df_ventas['region'].value_counts(ascending=False)
conteo_relativo = df_ventas['region'].value_counts(normalize=True, ascending=False) * 100

# Unimos ambas series en un DataFrame robusto
df_logistica = pd.concat([conteo_absoluto, conteo_relativo], axis=1)
df_logistica.columns = ['Nº Pedidos', 'Participación (%)']
df_logistica.index.name = 'Región'

# ==============================================================================
# 2. DEFINICIÓN DE CSS Y CAPA DE PRESENTACIÓN
# ==============================================================================
# Diseño "Clean" (Fondo blanco para no competir con las barras de progreso)
estilos_limpios = [
    {'selector': 'th', 'props': [('font-family', 'Arial'), ('font-weight', 'bold'), ('text-align', 'right'), ('background-color', 'white'), ('color', 'black')]},
    {'selector': 'td', 'props': [('background-color', 'white'), ('color', 'black')]}
]

# Color corporativo para las barras (Verde Menta)
COLOR_BARRA = '#9cd3bb'

# ==============================================================================
# 3. RENDERIZADO (Mini-Gráficos In-Cell)
# ==============================================================================
reporte_logistica_estilizado = (
    df_logistica.style
    .set_table_styles(estilos_limpios)
    .format({'Nº Pedidos': '{:,}', 'Participación (%)': '{:.2f}%'}) # Formato condicional mixto
    .bar(
        subset=['Participación (%)'], # La columna que dibujará la barra
        vmin=0, vmax=100,             # La escala (0 al 100%)
        color=COLOR_BARRA,            # El color de la barra
        height=70,                    # Altura de la barra (porcentaje de la celda)
        width=80                      # Ancho máximo de la barra
    )
)

# Despliegue del reporte
display(reporte_logistica_estilizado)
````

### 📅 Misión 4: Análisis de Estacionalidad Logística (Pivot Tables & Sticky CSS)

**Contexto del Negocio:**
El equipo logístico necesita prever picos de demanda. Transformaremos el historial transaccional en una matriz temporal (Tabla Dinámica) donde las filas son Departamentos y las columnas son Meses.

**Arquitectura del Código (Clean Code & CSS):**
1. **Ordenamiento Cronológico Base:** Al ordenar por `fecha_pedido` antes de extraer el string del mes (`%Y-%b`), garantizamos que al usar `sort=False` en el `pivot_table`, las columnas queden perfectamente ordenadas de enero de 2016 a diciembre de 2019.
2. **Corrección del *Scroll Overlap*:** El solapamiento al usar `.set_sticky()` ocurre por la transparencia HTML. Lo solucionamos inyectando la propiedad `background-color` (color sólido) y `z-index: 1` a los selectores `th.row_heading` y `.index_name`, asegurando que el índice flote opacamente sobre los datos.

````python
import pandas as pd

# ==============================================================================
# 1. PIPELINE DE TRANSFORMACIÓN (Tabla Dinámica Temporal)
# ==============================================================================
print("📦 Procesando Matriz de Estacionalidad Logística...")

# Habilitamos la vista amplia de Pandas para permitir el scroll
pd.set_option('display.max_columns', None)

df_estacionalidad = (
    df_ventas
    .sort_values('fecha_pedido')                                           # 1. Garantizamos el orden cronológico estricto
    .assign(Mes_Format=lambda x: x['fecha_pedido'].dt.strftime('%Y-%b'))   # 2. Extraemos Año-Mes (Ej: 2016-Jan)
    .pivot_table(
        index='departamento',                                              # Filas
        columns='Mes_Format',                                              # Columnas
        values='cantidad',                                                 # Datos a calcular
        aggfunc='sum',                                                     # Operación matemática
        sort=False                                                         # Evita que ordene alfabéticamente (rompería la cronología)
    )
    .rename_axis('Departamento', axis='index')                             # Limpieza de nombres
    .rename_axis('Periodo', axis='columns')
)

# ==============================================================================
# 2. UI/UX Y SOLUCIÓN DEL OVERLAP (Capa de Presentación)
# ==============================================================================
# CSS avanzado para solucionar el solapamiento de la columna congelada (Sticky)
estilos_matriz = [
    # Diseño de la esquina superior izquierda (Nombre del índice)
    {'selector': '.index_name', 'props': [('background-color', '#343a40'), ('color', 'white'), ('font-weight', 'bold'), ('z-index', '2')]},

    # Diseño de la columna congelada (Departamentos). Le damos fondo sólido claro para tapar los datos que pasan por debajo
    {'selector': 'th.row_heading', 'props': [('background-color', '#f8f9fa'), ('color', '#212529'), ('font-weight', 'bold'), ('z-index', '1')]},

    # Diseño de los meses (Encabezados de columna)
    {'selector': 'th.col_heading', 'props': [('text-align', 'center'), ('font-size', '0.9em')]},

    # Diseño del cuerpo numérico
    {'selector': 'td', 'props': [('text-align', 'center'), ('padding', '5px 15px')]}
]

reporte_logistica_estacional = (
    df_estacionalidad.style
    .format("{:,.0f}")                          # Formato de números enteros con separador de miles
    .set_sticky(axis='index')                   # Congelamos la primera columna para scroll horizontal
    .set_table_styles(estilos_matriz,overwrite=False)           # Inyectamos nuestra solución CSS
)

# Renderizado final
display(reporte_logistica_estacional)
````

### 📚 Teoría de Transformación de Datos: `pivot` vs `pivot_table`



En la manipulación de datos matriciales, a menudo necesitamos "girar" nuestro DataFrame para ver la información desde otra perspectiva. Pandas nos ofrece dos métodos que suenan igual pero tienen un comportamiento algorítmico completamente distinto.

**1. El Método Estructural (`pivot`):**
* **Función:** Simplemente reorganiza la forma de la tabla (Pasa de formato "Largo/Tidy" a "Ancho/Wide").
* **Limitación Crítica:** **NO hace cálculos matemáticos**. Si en tus datos originales hay dos filas que coinciden en el mismo Índice y Columna, el método `pivot` colapsará y arrojará un error (`ValueError: Index contains duplicate entries, cannot reshape`), porque no sabe qué valor poner en la nueva celda.

**2. El Método Analítico (`pivot_table`):**
* **Función:** Genera resúmenes estadísticos. Es el equivalente exacto a las "Tablas Dinámicas" de Excel.
* **Ventaja:** **SÍ hace cálculos matemáticos** a través del parámetro `aggfunc` (Suma, Media, Conteo, etc.). Si encuentra registros duplicados para una misma intersección, los suma (o promedia) y entrega un único valor limpio.

### 📈 Misión 5: Informe Ejecutivo de Desempeño (Cross-Tabulation)

**Contexto del Negocio:**
La Junta Directiva necesita una matriz de doble entrada (Cross-tab) que cruce el Canal de Ventas (B2B/B2C) con el Método de Envío, incluyendo Subtotales por categoría y un Gran Total General. Adicionalmente, se requiere destacar el canal/método que generó el mayor volumen de ingresos.

**Arquitectura del Código (Nivel Ejecutivo):**
1. **Cálculo Automático de Márgenes:** En lugar de hacer `.sum()` manual por filas y columnas, utilizamos `pd.crosstab(margins=True)`. Esto genera la matriz completa con sus totales en una sola transacción eficiente.
2. **CSS Condicional Dinámico (`apply`):** En lugar de crear un DataFrame booleano manual para inyectar clases (`set_td_classes`), usamos el método `.apply()` nativo del `Styler`. Creamos una función lógica que encuentra matemáticamente el valor máximo en la fila "B2C" (nuestro segmento estrella) y lo resalta automáticamente, sin importar cómo cambien los datos en el futuro.

````python
import pandas as pd
import numpy as np

# ==============================================================================
# 1. MATRIZ EJECUTIVA (Cross-Tabulation)
# ==============================================================================
print("🏢 Procesando Matriz Ejecutiva de Desempeño (Dinamismo Absoluto)...")

df_ejecutivo = pd.crosstab(
    index=df_ventas['segmento_cliente'],
    columns=df_ventas['modo_envio'],
    values=df_ventas['ventas'],
    aggfunc='sum',
    margins=True,
    margins_name='TOTAL'
).rename_axis(index=None, columns=None)

# ==============================================================================
# 2. ARQUITECTURA VISUAL (Estilo Bloque Oscuro)
# ==============================================================================
estilos_oscuros = [
    {'selector': 'td', 'props': [('font-family', 'Arial, sans-serif'), ('text-align', 'center'), ('padding', '10px')]},
    {'selector': 'th', 'props': [
        ('font-family', 'Arial, sans-serif'),
        ('font-weight', 'bold'),
        ('background-color', '#343a40'),
        ('color', '#ffffff'),
        ('text-align', 'center'),
        ('padding', '10px'),
        ('border', '1px solid #495057')
    ]},
    {'selector': 'tr:last-child', 'props': [('border-top', '3px solid #343a40'), ('font-weight', 'bold')]}
]

# ==============================================================================
# 3. LÓGICA DINÁMICA NATIVA (Sin Hardcoding)
# ==============================================================================
# Aislamos matemáticamente el "Core" de datos (Eliminamos la fila y columna 'TOTAL')
filas_core = df_ejecutivo.index.drop('TOTAL')
cols_core = df_ejecutivo.columns.drop('TOTAL')

# Definimos el estilo del ganador
CSS_GANADOR = 'background-color: #B86F5F; color: white; font-weight: bold; border-radius: 4px; box-shadow: inset 0 0 5px rgba(0,0,0,0.2);'

# ==============================================================================
# 4. RENDERIZADO DEL INFORME FINAL
# ==============================================================================
reporte_ejecutivo_final = (
    df_ejecutivo.style
    .format('${:,.0f}')
    .set_table_styles(estilos_oscuros)

    # 🚨 CLAVE: Buscamos el Máximo Global (axis=None) solo en la porción de datos reales
    .highlight_max(
        subset=(filas_core, cols_core),
        axis=None,
        props=CSS_GANADOR
    )
)

display(reporte_ejecutivo_final)
````

### 📐 Arquitectura de la Información: Tablas Formales y Científicas

**Contexto del Negocio y Auditoría:**
Mientras que los cuadros de mando interactivos (Dashboards) priorizan la agilidad visual, las **Tablas Formales** (o Académicas/Científicas) priorizan el rigor, la precisión y la auto-explicación. Son el estándar requerido para reportes financieros oficiales, auditorías externas y publicaciones rigurosas.



**Principios de Diseño Riguroso:**
Para que una tabla sea considerada "formal", debe sostenerse por sí misma (sin necesidad de leer el texto que la rodea) y minimizar las distracciones visuales:

1. **Alta Proporción Datos-Tinta (Simplicidad Visual):** Se minimiza el uso de tinta en elementos no ligados a los datos. Evitamos colores de fondo saturados y eliminamos por completo los bordes verticales. Solo se utilizan bordes horizontales sutiles para enmarcar los encabezados y los totales generales.
2. **Autonomía Semántica (Títulos y Unidades):** * **Título Descriptivo:** Debe responder *Qué, Dónde y Cuándo* (Ej. no uses "Ventas", usa "Ingresos Netos por Canal de Venta - Q4 2025").
   * **Unidades de Medida:** Las métricas deben especificar su unidad (`[USD]`, `[%]`, `[Unidades]`) en el encabezado o en una leyenda anexa para no ensuciar los números de cada celda.
3. **Integridad del Dato (Cero Celdas Vacías):** En auditoría y ciencia, un espacio en blanco genera ambigüedad grave (¿Es un cero? ¿Se perdió el dato durante el ETL? ¿No aplica?). Todo valor nulo debe ser formateado explícitamente usando `NaN`, `N/A`, `S/D` o un guion `-`.
4. **Paginación y Fragmentación:** Las sábanas de datos infinitas destruyen la legibilidad. Si un reporte es muy extenso, la mejor práctica es modularlo en tablas más pequeñas categorizadas por dimensiones lógicas.

### 🔬 Ejemplo Práctico: Renderizado de una Tabla Formal (Estilo Científico/APA)

**Arquitectura Visual:**
Para lograr el estándar académico y de auditoría, utilizamos las siguientes técnicas en el objeto `Styler`:
1. **`.hide(axis='index')`:** Ocultamos el índice numérico automático de Pandas, ya que no aporta valor semántico.
2. **`.set_caption()`:** Añadimos el título auto-explicativo que exige la norma formal.
3. **`na_rep='S/D'`:** Forzamos a que cualquier valor nulo (`NaN`) se renderice como "S/D" (Sin Datos), garantizando la integridad de lectura.
4. **CSS Minimalista:** Eliminamos todas las líneas verticales y solo dejamos un borde superior grueso, un borde sutil bajo los encabezados, y un borde de cierre al final de la tabla.

````python
import pandas as pd
import numpy as np

# ==============================================================================
# 1. DATOS DE MUESTRA (Con un valor nulo intencional para auditoría)
# ==============================================================================
datos_pyme = {
    'Región Estratégica': ['Buenos Aires', 'Centro', 'Litoral', 'Patagonia'],
    'Ingresos [USD]': [45000.50, 32000.00, np.nan, 28500.75],
    'Variación Anual [%]': [5.2, -1.5, 3.4, 8.1]
}
df_formal = pd.DataFrame(datos_pyme)

# ==============================================================================
# 2. DEFINICIÓN DEL ESTILO ACADÉMICO / FORMAL (Tinta Mínima)
# ==============================================================================
estilos_cientificos = [
    # General: Fondo blanco y texto negro
    {'selector': 'table, th, td', 'props': [('background-color', 'white'), ('color', 'black'), ('font-family', 'Times New Roman, serif')]},

    # Eliminamos todos los bordes por defecto y centramos el contenido
    {'selector': 'th, td', 'props': [('border', 'none'), ('padding', '8px 15px'), ('text-align', 'center')]},

    # Alineamos a la izquierda la primera columna (Textos/Categorías)
    {'selector': 'td:first-child, th:first-child', 'props': [('text-align', 'left')]},

    # Bordes horizontales para enmarcar el encabezado
    {'selector': 'thead th', 'props': [('border-top', '2px solid black'), ('border-bottom', '1px solid black'), ('font-weight', 'bold')]},

    # Borde inferior para cerrar la tabla
    {'selector': 'tbody tr:last-child td', 'props': [('border-bottom', '2px solid black')]},

    # Estilo del Título (Caption)
    {'selector': 'caption', 'props': [('caption-side', 'top'), ('text-align', 'left'), ('font-weight', 'bold'), ('color', 'black'), ('padding-bottom', '10px')]}
]

# ==============================================================================
# 3. CAPA DE PRESENTACIÓN (Styler)
# ==============================================================================
reporte_formal = (
    df_formal.style
    .hide(axis='index') # Escondemos el índice (0, 1, 2, 3) por estética formal
    .set_caption("Tabla 1. Rendimiento Financiero Regional y Variación Interanual (Q4 2025)")
    .format({
        'Ingresos [USD]': '{:,.2f}',
        'Variación Anual [%]': '{:.1f}'
    }, na_rep="S/D") # Manejo explícito de nulos (Sin Datos)
    .set_table_styles(estilos_cientificos)
)

# Renderizado
display(reporte_formal)
````

## 4. Estructuras de Datos Gerenciales (ETL Pre-Visualización)
No se puede estilizar lo que no está bien estructurado. Reglas para pivotar datos:

* ```pivot_table```: Agrupa y **calcula** matemáticas (Ej. Suma de ventas por mes y departamento). Equivalente a Tablas Dinámicas de Excel.

* ```crosstab```: Ideal para frecuencias o cruces de variables categóricas. Su superpoder es ```margins=True``` (calcula subtotales y Gran Total automáticamente).

* ```pivot```: Solo cambia la forma de la tabla (reshape). **Rompe** si hay datos duplicados en el cruce porque no sabe cómo sumarlos.

## 🧠 Buenas Prácticas (Reglas Senior)
1. **Separación de Capas:** Nunca mezcles cálculos matemáticos (.groupby, .sum()) con inyección visual (.style). Calcula primero, estiliza al final.

2. **Especificidad CSS:** El último método aplicado en el Styler sobrescribe al anterior. Aplica primero los mapas de calor (background_gradient) y luego los resaltados específicos (highlight_max).

3. **Accesibilidad (Modo Oscuro):** Nunca cambies el background-color de una celda sin forzar explícitamente el color (tipografía) para garantizar el contraste en cualquier pantalla.

4. **Cero Hardcoding:** Usa axis=None o funciones dinámicas en lugar de escribir nombres literales de celdas para evitar que el código se rompa cuando los datos cambien.
