### ⚙️ Optimización de Ingesta: Filtrado en la Lectura (`nrows` y `usecols`)

**Concepto de Negocio:**
Al importar grandes volúmenes de datos (ej. histórico de transacciones o padrón de clientes), cargar el archivo completo en memoria es ineficiente y riesgoso. Para análisis específicos (como la segmentación de los mejores clientes), solo necesitamos una fracción de la información.

**Estrategia de Solución:**
1. **`nrows` (Muestreo Rápido):** Permite leer solo las primeras $N$ filas. Es ideal para inspeccionar la estructura de un archivo masivo sin saturar la RAM.
2. **`usecols` (Selección de Atributos):** Permite indicar exactamente qué columnas cargar en memoria.
   * *Mejor Práctica (Clean Code):* Definir siempre las columnas por su **nombre exacto** mediante una lista, nunca por su índice numérico, para evitar la corrupción de datos si la estructura del archivo origen cambia en el futuro.


💡 Tip de Senior (Clean Code)
Cuando te enfrentes a un archivo CSV nuevo o a una tabla de base de datos que no creaste tú, nunca confíes ciegamente en la documentación o en lo que crees que dice. Antes de filtrar, siempre es buena práctica pedirle a Pandas que te liste los nombres exactos de las columnas con este comando corto:

```
print(df_muestra_clientes.columns.tolist())
```




````python
import pandas as pd

# 1. Constantes y Rutas
RUTA_ARCHIVO = '/content/drive/MyDrive/Pandas/superstore_data.csv'

# -------------------------------------------------------------------
# EJEMPLO A: Muestreo rápido de estructura (nrows)
# -------------------------------------------------------------------
CANTIDAD_FILAS_MUESTRA = 5

df_muestra_clientes = pd.read_csv(RUTA_ARCHIVO, nrows=CANTIDAD_FILAS_MUESTRA)
print("--- Muestra Rápida (Primeras 5 filas) ---")
display(df_muestra_clientes) # En Colab, display() renderiza mejor que print()


# -------------------------------------------------------------------
# EJEMPLO B: Ingesta optimizada para Análisis de Rendimiento (usecols)
# -------------------------------------------------------------------
# Definimos claramente el "contrato" de las columnas que nuestro algoritmo necesita
print(df_muestra_clientes.columns.tolist())

COLUMNAS_ANALISIS_CLIENTES = ['Id', 'Year_Birth', 'Income']

# Cargamos solo los atributos necesarios para la lógica de negocio
df_clientes_rfm = pd.read_csv(RUTA_ARCHIVO, usecols=COLUMNAS_ANALISIS_CLIENTES)

print("\n--- Datos Listos para Análisis (Filtrado por Columnas) ---")
display(df_clientes_rfm.head())
````

### 🛠️ Resolución de Errores de Ingesta: Codificación (Encoding)

**Concepto de Negocio:**
Al importar datos externos (ej. listados de ventas de sucursales o proveedores), es común que los archivos no vengan en el estándar actual (`UTF-8`), sino en formatos heredados de sistemas más antiguos (como `UTF-16` o `ISO-8859-1`). Si intentamos leerlos directamente, Pandas lanzará un `UnicodeDecodeError` al no saber cómo interpretar caracteres especiales.

**Estrategia de Solución:**
1. Utilizar la librería `chardet` para auditar el archivo.
2. Leer únicamente una **muestra** del archivo en formato binario para detectar el "alfabeto" (encoding) sin saturar la memoria RAM del servidor.
3. Inyectar el encoding detectado como parámetro en la función de lectura.

````python
import pandas as pd
import chardet

# 1. Definición de Constantes (Buenas prácticas para evitar "Hardcoding")
RUTA_ARCHIVO_VENTAS = '/content/drive/MyDrive/Pandas/superstore_data.csv'
TAMANO_MUESTRA_BYTES = 10000 # Solo leemos los primeros 10KB para no saturar la RAM

# 2. Detección dinámica del Encoding
with open(RUTA_ARCHIVO_VENTAS, 'rb') as archivo_crudo:
    # Leemos solo una porción representativa del archivo
    muestra_datos = archivo_crudo.read(TAMANO_MUESTRA_BYTES)
    resultado_auditoria = chardet.detect(muestra_datos)

# Extraemos los valores del diccionario resultante
encoding_detectado = resultado_auditoria['encoding']
nivel_confianza = resultado_auditoria['confidence']

print(f"🔍 Auditoría completada: Encoding '{encoding_detectado}' detectado con una confianza del {nivel_confianza:.2%}")

# 3. Ingesta de Datos con el parámetro dinámico
# Nota: En producción, podríamos agregar un 'if nivel_confianza > 0.8:' para mayor seguridad
df_ventas = pd.read_csv(RUTA_ARCHIVO_VENTAS, encoding=encoding_detectado)

# 4. Inspección de calidad
df_ventas.head()
````

### 🛡️ Ingesta Robusta: Normalización de Cabeceras (Clean Code)

**Concepto de Negocio:**
En los sistemas de gestión reales, los archivos externos suelen venir con inconsistencias tipográficas (ej. `Id`, `ID`, `  id `). Utilizar índices numéricos (`[0, 1, 4]`) para saltar este problema es peligroso, ya que si el proveedor cambia el orden de las columnas, el sistema ingerirá datos incorrectos silenciosamente.

**Estrategia de Solución:**
En lugar de adaptar nuestro código a los datos sucios, obligamos a los datos entrantes a pasar por un filtro de estandarización.
1. Definimos nuestro "Contrato de Datos" (las columnas que necesitamos) estrictamente en minúsculas.
2. Le pasamos una **función evaluadora** al parámetro `usecols` en lugar de una lista estática.
3. Pandas pasará el nombre de cada columna del CSV por nuestra función (que limpiará espacios y mayúsculas) y solo cargará la columna si el resultado coincide con nuestro contrato.

````python
import pandas as pd

# 1. Rutas
RUTA_ARCHIVO = '/content/drive/MyDrive/Pandas/superstore_data.csv'

# 2. Nuestro "Contrato Ideal" (Estandarizado en minúsculas)
COLUMNAS_ANALISIS_CLIENTES = ['id', 'year_birth', 'income']

# 3. Función Evaluadora / Normalizadora
def validar_columna(nombre_columna_original):
    """
    Toma el nombre de la columna del CSV, elimina espacios en los bordes
    y la convierte a minúsculas. Retorna True si coincide con nuestro contrato.
    """
    nombre_limpio = str(nombre_columna_original).strip().lower()
    return nombre_limpio in COLUMNAS_ANALISIS_CLIENTES

# 4. Lectura Robusta (Inyección de dependencia en usecols)
# Al pasar la función (sin paréntesis), Pandas la ejecuta internamente por cada columna.
df_clientes_rfm = pd.read_csv(RUTA_ARCHIVO, usecols=validar_columna)

print("\n--- Datos Listos para Análisis (Resistente a errores tipográficos) ---")
display(df_clientes_rfm.head())
````

### 🧠 ¿Cómo funciona el motor de `usecols` con funciones? (Under the Hood)

Cuando le pasamos una función a `usecols` en lugar de una lista, Pandas cambia su comportamiento predeterminado y actúa como un **guardia de seguridad interactivo**:

1. **Lectura de Cabecera:** Pandas lee únicamente la primera fila del CSV (los nombres de las columnas).
2. **Iteración Individual:** Toma el nombre de la primera columna y lo inyecta como argumento en nuestra función `validar_columna`.
3. **El Operador `in`:** La instrucción `return nombre_limpio in COLUMNAS_ANALISIS_CLIENTES` compara el resultado limpio contra *todos* los elementos de nuestra lista.
   * Si la lista tuviera valores repetidos (ej. `['id', 'id', 'income']`), Python simplemente devuelve `True` al encontrar la primera coincidencia. No genera error, pero es una redundancia innecesaria en el código (Clean Code dicta evitarlo).
4. **El Veredicto (`True` o `False`):** Si la función retorna `True`, Pandas anota esa columna en su memoria para descargar todos sus datos. Si retorna `False`, la descarta por completo.
5. **Repetición:** Repite el proceso con la segunda columna, la tercera, etc., hasta terminar la cabecera. Recién ahí, comienza a descargar las filas de las columnas aprobadas.

**⚠️ Casos Límite (Edge Cases) en Producción:**
¿Qué ocurre si el archivo del supermercado viene sucio y trae dos columnas casi idénticas, por ejemplo, `ID` y `Id`?
* Nuestra función limpiadora convertirá ambas a `'id'`.
* Ambas evaluarán como `True` al compararse con nuestra lista.
* **El Resultado de Pandas:** Cargará *ambas* columnas en la memoria. Sin embargo, como las bases de datos relacionales y los DataFrames no permiten nombres de columnas duplicados, Pandas aplicará una regla de auto-corrección silenciosa: renombrará la segunda columna como `id.1`. Esto es un riesgo de negocio, ya que el sistema podría procesar identificadores incorrectos.

### 💾 Exportación de Datos: Guardando Resultados (`to_csv`)

**Concepto de Negocio:**
Una vez que hemos limpiado y filtrado los datos (ej. seleccionando solo los clientes de alto rendimiento), necesitamos persistir esa información en un nuevo archivo para compartirlo con otras áreas o conectarlo a otro sistema.

**La Trampa del Índice (`Unnamed: 0`):**
Por defecto, Pandas exporta el "número de fila" interno (índice numérico) como si fuera una columna más de datos. Al volver a leer el archivo, esto crea una columna de "basura" llamada `Unnamed: 0`.

**Estrategia de Solución (Clean Code):**
* Utilizar siempre el parámetro `index=False` en la función `to_csv()`, a menos que el índice contenga información vital del negocio (ej. una fecha o un ID que nosotros mismos hayamos asignado como índice).
* Definir la ruta de exportación absoluta hacia nuestro Google Drive para garantizar que el archivo no se borre al cerrar la sesión de Colab.

````python
import pandas as pd

# 1. Definición de Rutas (Garantizando el guardado en tu Drive)
# Usamos un nombre descriptivo para el archivo final
RUTA_EXPORTACION = '/content/drive/MyDrive/Pandas/clientes_alta_rentabilidad.csv'

# 2. Exportación de Datos
# Asumimos que df_clientes_rfm es el DataFrame limpio del paso anterior.
# index=False bloquea la exportación del número de fila interno de Pandas.
df_clientes_rfm.to_csv(RUTA_EXPORTACION, index=False)

print(f"✅ Archivo exportado exitosamente de forma permanente en: {RUTA_EXPORTACION}")

# 3. Auditoría Rápida (Verificamos que se guardó sin la columna 'Unnamed: 0')
df_verificacion = pd.read_csv(RUTA_EXPORTACION)
print("\n--- Vista previa de la base de datos exportada ---")
display(df_verificacion.head())
````

### 🏋️‍♂️ Desafío Práctico: Ingesta de Datos Complejos (DATASUS)

**Contexto del Negocio:**
Disponemos de un archivo en formato CSV obtenido de la página *Información de Salud (TABNET) - DATASUS*. Este archivo contiene el registro de gastos hospitalarios públicos de cada una de las Unidades de la Federación Brasileña (enero 2008 - marzo 2021).

**Objetivo:**
Ingestar el archivo utilizando la función `pd.read_csv` de Pandas, limpiando la "basura" (metadatos) que viene incrustada en el archivo original antes de que llegue a nuestro DataFrame.

**Requisitos Técnicos (Parámetros a configurar):**
1. **Separador:** Inspeccionar si el archivo utiliza coma (`,`) o punto y coma (`;`).
2. **Codificación (Encoding):** El sistema origen utiliza el alfabeto `ISO-8859-1`.
3. **Omisión de Cabeceras (Skip Rows):** Ignorar las primeras 3 líneas del archivo, ya que el encabezado tabular real comienza recién en la cuarta línea.
4. **Omisión de Pie de Página (Skip Footer):** Ignorar las últimas 9 líneas, ya que son notas al pie sobre la fuente de los datos y romperían la estructura de la base de datos.
5. **Motor de Procesamiento:** Para que Pandas pueda aplicar la omisión del pie de página de manera segura, es estrictamente necesario habilitar el motor nativo agregando el parámetro `engine='python'`.

````python
import pandas as pd
import chardet

# 1. Definición de Rutas y Constantes
RUTA_ARCHIVO_SALUD = '/content/drive/MyDrive/Pandas/datos_sus.csv'
TAMANO_MUESTRA_BYTES = 20000

# 2. Detección dinámica del Encoding
with open(RUTA_ARCHIVO_SALUD, 'rb') as archivo_crudo:
    muestra_datos = archivo_crudo.read(TAMANO_MUESTRA_BYTES)
    resultado_auditoria = chardet.detect(muestra_datos)

encoding_detectado = resultado_auditoria['encoding']
print(f"🔍 Auditoría completada: Encoding '{encoding_detectado}' detectado.")

# 3. Ingesta de Datos Estricta (Inyección de Parámetros de Limpieza)
df_salud = pd.read_csv(
    RUTA_ARCHIVO_SALUD,
    encoding=encoding_detectado,
    sep=';',              # Regla 1: El separador es punto y coma
    skiprows=3,           # Regla 2: Saltamos los 3 primeros renglones de basura
    skipfooter=9,         # Regla 3: Saltamos los 9 últimos renglones de notas al pie
    engine='python'       # Regla 4: Obligatorio para que skipfooter funcione
)

# 4. Inspección de calidad
print("\n--- Base de Datos DATASUS (Limpiada) ---")
display(df_salud.head())
````
