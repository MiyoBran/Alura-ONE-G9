##### Ventajas array NumPy

Existen varias ventajas en el uso de arrays Numpy en lugar de listas regulares de Python y aquí tienes algunas de ellas:

1. Eficiencia de procesamiento: Las operaciones matemáticas en los arrays Numpy son mucho más rápidas que en las listas regulares, ya que Numpy está optimizado para trabajar con conjuntos de datos homogéneos y libera memoria de la computadora de manera rápida.

2. Facilidad de uso: Las operaciones matemáticas en los arrays Numpy se expresan de manera mucho más clara y concisa que en las listas regulares, lo que hace que el código sea más fácil de leer y mantener.

3. Integración con otras bibliotecas: Numpy es una de las bibliotecas más utilizadas en ciencia de datos y aprendizaje automático. Muchas otras bibliotecas, como Pandas y Matplotlib, están diseñadas para trabajar directamente con arrays Numpy.

````python
import numpy as np

# Simulamos el problema
krasnodar_copy = Krasnodar.copy()

# 1. Auditoría de Calidad
cantidad_nans = np.sum(np.isnan(krasnodar_copy))

if cantidad_nans > 0:
    print(f"⚠️ ALERTA: Se detectaron {cantidad_nans} valores faltantes.")

    # 2. Localizar el índice del problema (Sin buscar a ojo)
    indices_nan = np.argwhere(np.isnan(krasnodar_copy))
    print(f"Índices afectados: {indices_nan.flatten()}")

    # 3. Imputación Dinámica (Clean Code)
    # Iteramos por cada error encontrado en lugar de poner [4] a mano
    for indice in indices_nan.flatten():
        # Verificamos bordes para no romper el array
        if 0 < indice < len(krasnodar_copy) - 1:
            valor_anterior = krasnodar_copy[indice - 1]
            valor_siguiente = krasnodar_copy[indice + 1]

            # Lógica de Interpolación Lineal
            promedio = (valor_anterior + valor_siguiente) / 2
            krasnodar_copy[indice] = promedio
            print(f"✅ Reparado índice {indice}: {promedio:.2f}")
        else:
            print(f"❌ No se puede interpolar en los bordes automáticamente.")

else:
    print("✅ Calidad de datos al 100%.")
````

#### **Clean Code** : regresion lineal manual

````python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. DEFINICIÓN DEL MODELO (Lógica de Negocio) ---
def calcular_pronostico_lineal(x: np.ndarray, tasa_crecimiento: float, precio_base: float) -> np.ndarray:
    """
    Implementa y = ax + b
    x: Array de tiempo (meses)
    tasa_crecimiento (a): Cuánto aumenta el precio por mes.
    precio_base (b): Precio inicial estimado.
    """
    return (tasa_crecimiento * x) + precio_base

def calcular_error_modelo(datos_reales: np.ndarray, datos_predichos: np.ndarray) -> float:
    """
    Calcula la distancia Euclidiana (Norma L2) entre la realidad y la ficción.
    Usa np.linalg.norm que está optimizado en C (mucho más rápido que un bucle).
    """
    return np.linalg.norm(datos_reales - datos_predichos)

# --- 2. CONFIGURACIÓN DE DATOS (Input) ---
# Asumimos que 'fechas' y 'Moscu' ya existen de los pasos anteriores
# fechas = np.arange(1, 88)
# Moscu = precios[:, 0]

# --- 3. SIMULACIÓN DE ESCENARIOS (Lo que hizo el instructor, pero estructurado) ---

# Escenario A: Inflación muy alta
tasa_A = 1.5
base_A = 78.0 # Precio inicial aproximado
prediccion_A = calcular_pronostico_lineal(fechas, tasa_A, base_A)
error_A = calcular_error_modelo(Moscu, prediccion_A)

# Escenario B: Inflación moderada (Ajuste)
tasa_B = 0.46 # Probando un valor más fino
base_B = 78.0
prediccion_B = calcular_pronostico_lineal(fechas, tasa_B, base_B)
error_B = calcular_error_modelo(Moscu, prediccion_B)

# --- 4. REPORTE DE RESULTADOS ---
print(f"Error del Modelo A (a={tasa_A}): {error_A:.2f}")
print(f"Error del Modelo B (a={tasa_B}): {error_B:.2f}")

if error_B < error_A:
    print("✅ El Modelo B se ajusta mejor a la realidad.")
else:
    print("✅ El Modelo A es mejor.")

# Visualización rápida para comparar
plt.plot(fechas, Moscu, label="Realidad (Moscu)")
plt.plot(fechas, prediccion_A, linestyle='--', label=f"Modelo A (Error: {int(error_A)})")
plt.plot(fechas, prediccion_B, linestyle='--', label=f"Modelo B (Error: {int(error_B)})")
plt.legend()
plt.show()
````

````python
import numpy as np

def calcular_regresion_lineal_manual(x: np.ndarray, y: np.ndarray) -> tuple[float, float]:
    """
    Calcula la pendiente (a) y la intersección (b) usando el método de Mínimos Cuadrados.

    Fórmulas:
    a = (n * sum(xy) - sum(x) * sum(y)) / (n * sum(x^2) - (sum(x))^2)
    b = promedio(y) - a * promedio(x)

    Args:
        x: Array de variable independiente (ej: Tiempo/Meses)
        y: Array de variable dependiente (ej: Precios)

    Returns:
        Tuple (pendiente, interseccion)
    """
    # 1. Validaciones de Arquitectura (Evitar crashes)
    if x.shape != y.shape:
        raise ValueError("Los arrays X e Y deben tener la misma dimensión.")

    n = x.shape[0] # Número de observaciones

    # 2. Pre-cálculo de componentes (Vectorización NumPy = Velocidad)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_xy = np.sum(x * y)      # Multiplicación elemento a elemento y luego suma
    sum_x_cuadrado = np.sum(x ** 2)

    # 3. Cálculo de la Pendiente (a)
    numerador = (n * sum_xy) - (sum_x * sum_y)
    denominador = (n * sum_x_cuadrado) - (sum_x ** 2)

    # Manejo de borde: Si el denominador es 0, X es constante (línea vertical)
    if denominador == 0:
        raise ValueError("No se puede calcular la regresión: La variable X no tiene variación.")

    a_pendiente = numerador / denominador

    # 4. Cálculo de la Intersección (b)
    # Usamos la fórmula simplificada basada en promedios
    promedio_x = np.mean(x)
    promedio_y = np.mean(y)

    b_interseccion = promedio_y - (a_pendiente * promedio_x)

    return a_pendiente, b_interseccion

# --- EJECUCIÓN CON TUS DATOS ---

# Asumimos que ya tienes 'fechas' y 'Moscu' cargados
try:
    a_real, b_real = calcular_regresion_lineal_manual(fechas, Moscu)

    print(f"✅ Modelo Matemático Exacto Generado:")
    print(f"   -> Pendiente (Inflación Mensual): {a_real:.4f} rublos/mes")
    print(f"   -> Intersección (Precio Base):    {b_real:.4f} rublos")

    # Cálculo del Error (Norma) para comparar con el "ojímetro" del video anterior
    y_predicho = (a_real * fechas) + b_real
    error_final = np.linalg.norm(Moscu - y_predicho)
    print(f"   -> Error Total (Norma): {error_final:.2f}")

except Exception as e:
    print(f"❌ Error: {e}")
````

### Clean Code: preasignacion de memoria
1. El Error Arquitectónico: np.append dentro de un for
    El instructor hizo esto:


    ```python
    norma = np.array([])
    for i in range(100):
        # ... calculo ...
        norma = np.append(norma, calculo)  # <--- ¡CUELLO DE BOTELLA!
    ```
    Por qué esto destruye el rendimiento:
    Los arrays de NumPy son inmutables en tamaño. No pueden crecer dinámicamente como las listas de Java (ArrayList) o Python (list).
    Cada vez que llamas a np.append, NumPy hace esto internamente:

    Crea un NUEVO array en memoria con tamaño N+1.

    COPIA todos los datos viejos al nuevo lugar.

    Agrega el dato nuevo.

    Borra el array viejo.

    Si haces esto 100 veces, no pasa nada. Si lo haces 1 millón de veces, estás copiando y borrando millones de datos innecesariamente. Es como construir una pared de ladrillos derrumbando la pared entera cada vez que pones un ladrillo nuevo.

2. La Solución Clean Code: Pre-asignación de Memoria
En Ingeniería de Datos, siempre reservamos la memoria antes de usarla. Si sabemos que vamos a hacer 100 simulaciones, creamos un array de ceros de tamaño 100 y luego lo rellenamos.

3. Implementación Robusta (Simulación de Escenarios)
Aquí tienes la versión refactorizada. Aplica Pre-asignación y Semántica de Negocio.





````python
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURACIÓN DE LA SIMULACIÓN ---
CANTIDAD_SIMULACIONES = 100
RANGO_PENDIENTE_MIN = 0.1
RANGO_PENDIENTE_MAX = 0.9

# Datos base (Asumiendo que ya existen 'fechas' y 'Moscu')
# fechas = ...
# Moscu = ...
# b = 79.64 (Intersección fija para este experimento)

# 1. GENERACIÓN DE ESCENARIOS (Inputs Aleatorios)
# Generamos 100 posibles tasas de inflación (pendientes) de una sola vez
pendientes_simuladas = np.random.uniform(
    RANGO_PENDIENTE_MIN,
    RANGO_PENDIENTE_MAX,
    CANTIDAD_SIMULACIONES
)

# 2. RESERVA DE MEMORIA (Clean Code: Pre-allocation)
# Creamos un contenedor vacío de tamaño fijo. Mucho más rápido que append.
errores_norma = np.zeros(CANTIDAD_SIMULACIONES)

# 3. MOTOR DE CÁLCULO (Iteración Optimizada)
for i in range(CANTIDAD_SIMULACIONES):
    # Extraemos la pendiente candidata
    pendiente_actual = pendientes_simuladas[i]

    # Calculamos la predicción con esa pendiente
    y_simulado = (pendiente_actual * fechas) + b

    # Calculamos el error (Norma) y lo guardamos en la posición reservada
    errores_norma[i] = np.linalg.norm(Moscu - y_simulado)

# --- ANÁLISIS DE RESULTADOS ---
# Buscamos cuál fue el menor error obtenido
mejor_error_idx = np.argmin(errores_norma) # Índice del valor más bajo
mejor_pendiente = pendientes_simuladas[mejor_error_idx]
mejor_error = errores_norma[mejor_error_idx]

print(f"--- RESULTADO DE LA SIMULACIÓN DE MONTECARLO ---")
print(f"Simulaciones ejecutadas: {CANTIDAD_SIMULACIONES}")
print(f"Mejor pendiente encontrada: {mejor_pendiente:.4f}")
print(f"Menor error logrado: {mejor_error:.2f}")

# (Opcional) Visualizamos cómo varían los errores
plt.plot(pendientes_simuladas, errores_norma, 'o', alpha=0.5)
plt.title("Búsqueda del Mejor Ajuste (Fuerza Bruta)")
plt.xlabel("Pendiente Probada")
plt.ylabel("Error (Norma)")
plt.show()
````

2. Clean Code & Arquitectura: "Global State" vs. "Instancia"

    El video enseña la forma "clásica" (y antigua) de NumPy:

````python
np.random.seed(99) # <--- Estado Global
np.random.random()
````

La Forma Senior (Modern NumPy - Recomendada):
Desde 2019 (NumPy 1.17+), la recomendación oficial es crear una Instancia de Generador (Generator Instance). Es Orientado a Objetos, seguro y encapsulado.

````python
# Clean Code: Encapsulamiento
rng = np.random.default_rng(seed=99) # Creamos un objeto generador "privado"
numeros = rng.uniform(0.1, 0.9, 10) # Usamos ese objeto
numeros
````

````python
import numpy as np

# --- CONFIGURACIÓN DE REPRODUCIBILIDAD ---
SEMILLA_AUDITORIA = 84  # La clave para repetir el experimento
rng = np.random.default_rng(seed=SEMILLA_AUDITORIA) # Generador aislado

# --- CONFIGURACIÓN DE NEGOCIO ---
CANTIDAD_SIMULACIONES = 1000 # Aumentamos a 1000 para más precisión
# Asumimos que 'fechas' y 'Moscu' existen, y 'b' (intersección) es 79.64
# Si no, defínelos brevemente antes o usa los del ejercicio anterior.
b_base = 79.64

# 1. GENERACIÓN DETERMINISTA
# Nota: Usamos 'rng' en lugar de 'np.random'
pendientes_simuladas = rng.uniform(0.1, 0.9, CANTIDAD_SIMULACIONES)

# 2. CÁLCULO VECTORIZADO (Sin bucles for - Nivel Senior)
# Explicación: NumPy permite restar una Matriz (Simulaciones) menos un Vector (Realidad)
# Esto crea una matriz de errores (1000 filas x 87 meses).
# Es avanzado, pero si te animas, es 100 veces más rápido que el bucle 'for'.
# Si prefieres el bucle 'for' por claridad, úsalo, pero con la semilla fija.

errores_norma = np.zeros(CANTIDAD_SIMULACIONES)

for i in range(CANTIDAD_SIMULACIONES):
    pendiente = pendientes_simuladas[i]
    y_simulado = (pendiente * fechas) + b_base
    errores_norma[i] = np.linalg.norm(Moscu - y_simulado)

# 3. RESULTADOS
mejor_idx = np.argmin(errores_norma)
mejor_pendiente = pendientes_simuladas[mejor_idx]
mejor_error = errores_norma[mejor_idx]

print(f"--- AUDITORÍA DE MODELO (Seed={SEMILLA_AUDITORIA}) ---")
print(f"Mejor Pendiente: {mejor_pendiente:.6f}") # Debería ser fija siempre
print(f"Error Mínimo:    {mejor_error:.6f}")
````
