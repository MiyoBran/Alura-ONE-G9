## 2Ô∏è‚É£ Crear Datasets de Ejemplo

Para este notebook, trabajaremos con dos datasets simulados que representan casos comunes en Data Science:

```python
# Dataset 1: Ventas Mensuales
np.random.seed(42)
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
ventas = np.array([15000, 18500, 16200, 22000, 25000, 28000])
gastos = np.array([8000, 9200, 8500, 11000, 12500, 14000])

# Crear DataFrame para mejor manejo de datos
df_ventas = pd.DataFrame({
    'mes': meses,
    'ventas': ventas,
    'gastos': gastos,
    'ganancia': ventas - gastos
})

print("üìä Dataset 1: Ventas Mensuales")
print(df_ventas)
print("\n" + "="*50 + "\n")

# Dataset 2: Temperatura por Hora
horas = np.arange(0, 24)
temperatura = np.array([15, 14, 13, 12, 12, 13, 15, 18, 22, 25, 28, 30,
                        31, 30, 29, 28, 26, 24, 22, 20, 18, 17, 16, 15])
humedad = np.array([85, 88, 90, 92, 91, 89, 80, 70, 60, 50, 45, 42,
                    40, 42, 45, 48, 55, 65, 75, 80, 82, 83, 84, 82])

df_clima = pd.DataFrame({
    'hora': horas,
    'temperatura': temperatura,
    'humedad': humedad
})

print("üå°Ô∏è Dataset 2: Datos de Clima")
print(df_clima.head(10))
```

## 3Ô∏è‚É£ Etiquetar Ejes: `xlabel()` y `ylabel()`

### ¬øPor qu√© es importante?
Las etiquetas de ejes son **fundamentales** para que cualquiera entienda qu√© mide cada eje sin necesidad de contexto adicional. Deben incluir:
- ‚úÖ Descripci√≥n clara del dato
- ‚úÖ Unidad de medida (USD, ¬∞C, %, etc.)
- ‚úÖ Formato consistente y legible

### Ejemplo 1: Gr√°fico Simple de Ventas

```python
# Crear figura con dimensiones adecuadas
plt.figure(figsize=(10, 6))

# Plotear datos
plt.plot(df_ventas['mes'], df_ventas['ventas'], 
         marker='o', linewidth=2.5, markersize=8, color=colores['primary'], label='Ventas')
plt.plot(df_ventas['mes'], df_ventas['gastos'], 
         marker='s', linewidth=2.5, markersize=8, color=colores['accent'], label='Gastos')

# ‚úÖ BUENA PR√ÅCTICA: Labels descriptivos con unidades de medida
plt.xlabel('Mes del A√±o', fontsize=12, fontweight='bold', color=colores['secondary'])
plt.ylabel('Monto (USD)', fontsize=12, fontweight='bold', color=colores['secondary'])

# Agregar t√≠tulo
plt.title('Comparaci√≥n de Ventas vs Gastos Mensuales', fontsize=14, fontweight='bold')

# Mejorar legibilidad
plt.legend(loc='upper left', fontsize=10, framealpha=0.95)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print("‚úÖ Se agregaron labels con:")
print("   - Descripci√≥n clara (Mes del A√±o, Monto)")
print("   - Unidad de medida (USD)")
print("   - Formato consistente (fontsize y fontweight)")
```

## 4Ô∏è‚É£ Personalizar Marcas de Ejes: `xticks()` y `yticks()`

### ¬øPara qu√© sirve?
Las marcas (ticks) son los valores mostrados en los ejes. Personalizarlas permite:
- ‚úÖ Mejorar legibilidad
- ‚úÖ Mostrar solo valores relevantes
- ‚úÖ Convertir formatos (decimales ‚Üí porcentajes)
- ‚úÖ Rotar etiquetas si es necesario

### Ejemplo 1: Datos de Temperatura (Ejes X mejorados)

```python
plt.figure(figsize=(12, 6))

# Graficar temperatura y humedad
ax1 = plt.gca()
color_temp = colores['primary']
ax1.plot(df_clima['hora'], df_clima['temperatura'], color=color_temp, marker='o', 
         linewidth=2.5, label='Temperatura (¬∞C)')
ax1.set_xlabel('Hora del D√≠a', fontsize=11, fontweight='bold')
ax1.set_ylabel('Temperatura (¬∞C)', fontsize=11, fontweight='bold', color=color_temp)
ax1.tick_params(axis='y', labelcolor=color_temp)

# ‚úÖ BUENA PR√ÅCTICA: Mostrar solo horas clave (cada 3 horas)
horas_etiquetas = np.arange(0, 24, 3)
etiquetas_formateadas = [f'{h:02d}:00' for h in horas_etiquetas]
ax1.set_xticks(horas_etiquetas)
ax1.set_xticklabels(etiquetas_formateadas, fontsize=10, rotation=45)

# Eje Y personalizado
yticks_temp = np.arange(10, 35, 5)
ax1.set_yticks(yticks_temp)
ax1.set_yticklabels([f'{t}¬∞' for t in yticks_temp], fontsize=10)

plt.title('Variaci√≥n de Temperatura - 24 Horas', fontsize=13, fontweight='bold', pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print("‚úÖ xticks() personalizado:")
print(f"   - Mostrar solo horas clave: {list(horas_etiquetas)}")
print(f"   - Formato: HH:MM")
print(f"   - Rotaci√≥n: 45¬∞")
print("\n‚úÖ yticks() personalizado:")
print(f"   - Valores: {list(yticks_temp)}")
print(f"   - Formato: T¬∞")
```

## 5Ô∏è‚É£ A√±adir Texto: `text()`

### Caracter√≠sticas
La funci√≥n `text()` permite escribir texto en posiciones espec√≠ficas del gr√°fico para:
- ‚úÖ Destacar valores clave
- ‚úÖ A√±adir anotaciones simples
- ‚úÖ Mostrar resultados o conclusiones
- ‚úÖ Mejorar la interpretabilidad

### Ejemplo: Ganancias Mensuales con Valores Destacados

```python
plt.figure(figsize=(10, 6))

# Crear gr√°fico de barras con ganancias
barras = plt.bar(df_ventas['mes'], df_ventas['ganancia'], 
                 color=colores['accent'], edgecolor='black', linewidth=1.5, alpha=0.8)

# ‚úÖ BUENA PR√ÅCTICA: A√±adir valor encima de cada barra
for i, (mes, ganancia) in enumerate(zip(df_ventas['mes'], df_ventas['ganancia'])):
    plt.text(i, ganancia + 300, f'${ganancia:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold',
             color=colores['secondary'])

# Resaltar ganancia m√°xima
max_idx = df_ventas['ganancia'].idxmax()
max_valor = df_ventas['ganancia'].max()

plt.text(max_idx, max_valor + 2000, f'‚≠ê Max: ${max_valor:,.0f}',
         ha='center', fontsize=11, fontweight='bold', color='red',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.6))

# A√±adir referencia de promedio
promedio = df_ventas['ganancia'].mean()
plt.axhline(y=promedio, color='red', linestyle='--', linewidth=2, alpha=0.7)
plt.text(5.5, promedio + 500, f'Promedio: ${promedio:,.0f}',
         ha='right', fontsize=10, fontweight='bold', 
         bbox=dict(boxstyle='round', facecolor='#ff9999', alpha=0.5))

plt.xlabel('Mes', fontsize=11, fontweight='bold')
plt.ylabel('Ganancia (USD)', fontsize=11, fontweight='bold')
plt.title('Ganancia Mensual con Anotaciones', fontsize=13, fontweight='bold')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("‚úÖ Se us√≥ text() para:")
print("   - Mostrar valor en cada barra")
print("   - Destacar el m√°ximo con backgroundColor")
print("   - Indicar el promedio con l√≠nea y etiqueta")
```

## 6Ô∏è‚É£ Crear Anotaciones: `annotate()`

### Ventajas de `annotate()` vs `text()`
- ‚úÖ Conecta autom√°ticamente el texto con un punto espec√≠fico
- ‚úÖ Permite usar flechas (`arrowprops`)
- ‚úÖ Mejor para se√±alar eventos o anomal√≠as
- ‚úÖ M√°s profesional y preciso

### Ejemplo: Detectar Anomal√≠as en Datos de Tr√°fico

```python
# Crear datos de tr√°fico simulados (con anomal√≠a)
np.random.seed(42)
dias = np.arange(1, 31)
trafico_base = np.random.normal(1200, 100, 30).astype(int)

# Introducir anomal√≠a (ca√≠da por mantenimiento)
trafico_base[8:12] = np.array([800, 750, 770, 800])

# Crear pico importante (campa√±a viral)
trafico_base[16:18] = np.array([2500, 2200])

plt.figure(figsize=(12, 6))
plt.plot(dias, trafico_base, marker='o', linewidth=2, color=colores['primary'], markersize=6)

# ‚úÖ BUENA PR√ÅCTICA: Anotar anomal√≠as con flechas
# 1. Se√±alar ca√≠da de tr√°fico
plt.annotate('‚ö†Ô∏è Mantenimiento del servidor',
             xy=(10, 775), xytext=(13, 500),
             arrowprops=dict(arrowstyle='->', color='red', lw=2, connectionstyle='arc3,rad=0.3'),
             fontsize=10, fontweight='bold', color='red',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='#ffcccc', alpha=0.8))

# 2. Se√±alar pico de tr√°fico
plt.annotate('üìà Campa√±a viral exitosa',
             xy=(17, 2500), xytext=(20, 2800),
             arrowprops=dict(arrowstyle='->', color='green', lw=2, connectionstyle='arc3,rad=-0.3'),
             fontsize=10, fontweight='bold', color='green',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='#ccffcc', alpha=0.8))

plt.xlabel('D√≠a del Mes', fontsize=11, fontweight='bold')
plt.ylabel('Sesiones', fontsize=11, fontweight='bold')
plt.title('An√°lisis de Tr√°fico Web con Anomal√≠as Detectadas', fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print("‚úÖ Se us√≥ annotate() para:")
print("   - Conectar texto con punto espec√≠fico (xy)")
print("   - Mostrar flecha (arrowprops)")
print("   - Proporcionar contexto de anomal√≠as")
```

## 7Ô∏è‚É£ T√≠tulo y Leyenda: `title()` y `legend()`

### Importancia
- **Title()**: Proporciona contexto general del gr√°fico
- **Legend()**: Diferencia entre m√∫ltiples series de datos

### Buenas Pr√°cticas
- ‚úÖ T√≠tulos descriptivos y concisos
- ‚úÖ Leyenda en posici√≥n que no oculte datos
- ‚úÖ Usar `framealpha` para mejor legibilidad

### Ejemplo: M√∫ltiples Series con Legendas

```python
plt.figure(figsize=(11, 6))

# Crear dos series de datos (Regi√≥n Norte y Regi√≥n Sur)
region_norte = df_ventas['ventas'].values
region_sur = df_ventas['ventas'].values * 0.75 + np.random.normal(0, 1000, len(df_ventas))

# Plotear ambas series
plt.plot(df_ventas['mes'], region_norte, marker='o', linewidth=2.5, markersize=8,
         label='Regi√≥n Norte', color=colores['primary'])
plt.plot(df_ventas['mes'], region_sur, marker='s', linewidth=2.5, markersize=8,
         label='Regi√≥n Sur', color=colores['accent'])

# ‚úÖ BUENA PR√ÅCTICA: T√≠tulo principal y subt√≠tulo
plt.title('Comparaci√≥n de Ventas por Regi√≥n', fontsize=14, fontweight='bold', pad=20)

# Subtitle (usando text)
plt.text(0.5, 0.95, 'An√°lisis del Primer Semestre 2024', 
         transform=plt.gca().transAxes, ha='center', 
         fontsize=10, style='italic', color='gray')

# ‚úÖ BUENA PR√ÅCTICA: Leyenda posicionada inteligentemente
plt.legend(loc='upper left', fontsize=11, framealpha=0.95, shadow=True, 
          title='Regiones', title_fontsize=11)

plt.xlabel('Mes', fontsize=11, fontweight='bold')
plt.ylabel('Ventas (USD)', fontsize=11, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

print("‚úÖ Elementos agregados:")
print("   - Title principal: 'Comparaci√≥n de Ventas por Regi√≥n'")
print("   - Subt√≠tulo: 'An√°lisis del Primer Semestre 2024'")
print("   - Legend con dos entradas y t√≠tulo")
print("   - Posici√≥n: upper left (√°rea sin datos)")
```

## 8Ô∏è‚É£ Personalizar Grilla: `grid()`

### ¬øCu√°ndo usar grid()?
- ‚úÖ Cuando necesitas leer valores precisos
- ‚úÖ Para mejorar claridad en scatter plots
- ‚úÖ En gr√°ficos cient√≠ficos o t√©cnicos
- ‚ùå NO sobrecargar visualizaciones

### Par√°metros Clave
- `alpha`: Transparencia (0-1, recomendado 0.2-0.4)
- `linestyle`: Tipo de l√≠nea ('--', ':', '-.')
- `linewidth`: Grosor de l√≠nea

```python
# Crear figura con 2 subplots para comparar grid vs sin grid
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Datos
horas_reducidas = np.arange(0, 24, 2)
temp_reducida = [15, 13, 15, 20, 25, 29, 30, 28, 25, 22, 19, 17]

# Subplot 1: Sin grid
ax1.plot(horas_reducidas, temp_reducida, marker='o', linewidth=2, color=colores['primary'], markersize=8)
ax1.set_xlabel('Hora', fontsize=11, fontweight='bold')
ax1.set_ylabel('Temperatura (¬∞C)', fontsize=11, fontweight='bold')
ax1.set_title('SIN Grid', fontsize=12, fontweight='bold')
ax1.set_xticks(horas_reducidas)

# Subplot 2: Con grid personalizado
ax2.plot(horas_reducidas, temp_reducida, marker='o', linewidth=2, color=colores['accent'], markersize=8)
ax2.set_xlabel('Hora', fontsize=11, fontweight='bold')
ax2.set_ylabel('Temperatura (¬∞C)', fontsize=11, fontweight='bold')
ax2.set_title('CON Grid (Sutil)', fontsize=12, fontweight='bold')
ax2.set_xticks(horas_reducidas)

# ‚úÖ BUENA PR√ÅCTICA: Grid moderado
ax2.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)

# A√±adir l√≠nea horizontal de referencia
ax2.axhline(y=22, color='red', linestyle=':', linewidth=1.5, alpha=0.5, label='Promedio')
ax2.legend(fontsize=10)

plt.suptitle('Comparaci√≥n: Grid vs Sin Grid', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("‚úÖ Grid personalizado:")
print("   - alpha=0.3 (transparencia moderada)")
print("   - linestyle='--' (l√≠neas discontinuas)")
print("   - Mejora legibilidad sin saturar el gr√°fico")
```

## 9Ô∏è‚É£ Proyecto Completo: Dashboard Profesional

### Objetivo
Integrar TODOS los elementos aprendidos en una visualizaci√≥n profesional que demuestre buenas pr√°cticas para un proyecto de Data Science.

### Elementos Integrados
‚úÖ Labels descriptivos (xlabel, ylabel)
‚úÖ Ticks personalizados (xticks, yticks)
‚úÖ Anotaciones con texto (text)
‚úÖ Anotaciones con flechas (annotate)
‚úÖ T√≠tulo y leyenda
‚úÖ Grid personalizado
‚úÖ Colores coherentes y profesionales

```python
# PROYECTO COMPLETO: Dashboard de An√°lisis de Ventas Online
"""
Simulamos un proyecto real de Data Science donde analizamos m√©tricas 
de una tienda online y sus tendencias mensuales.
"""

# Datos del proyecto
productos = {
    'Laptop': {'ventas': 450, 'ingresos': 45000, 'costo': 35000},
    'Mouse': {'ventas': 2800, 'ingresos': 1400, 'costo': 700},
    'Monitor': {'ventas': 320, 'ingresos': 9600, 'costo': 5000},
    'Teclado': {'ventas': 1900, 'ingresos': 950, 'costo': 500},
    'Webcam': {'ventas': 1500, 'ingresos': 900, 'costo': 400}
}

# Crear DataFrame
df_productos = pd.DataFrame(productos).T
df_productos['ganancia'] = df_productos['ingresos'] - df_productos['costo']

# Crear figura con 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ========== GR√ÅFICO 1: Ventas por Producto ==========
colores_productos = [colores['primary'], colores['secondary'], colores['accent'], 
                     colores['gray'], '#f59e0b']

barras1 = ax1.bar(df_productos.index, df_productos['ventas'], 
                  color=colores_productos, edgecolor='black', linewidth=1.5, alpha=0.85)

# ‚úÖ Agregar valores encima de barras
for i, (bar, ventas) in enumerate(zip(barras1, df_productos['ventas'])):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(ventas)} unid.',
             ha='center', va='bottom', fontsize=9, fontweight='bold')

# ‚úÖ Se√±alar producto m√°s vendido
max_idx = df_productos['ventas'].idxmax()
ax1.annotate('üèÜ M√°s Vendido',
             xy=(list(df_productos.index).index(max_idx), df_productos.loc[max_idx, 'ventas']),
             xytext=(list(df_productos.index).index(max_idx) - 0.5, df_productos.loc[max_idx, 'ventas'] + 600),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=10, fontweight='bold', color='red',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7))

ax1.set_ylabel('Unidades Vendidas', fontsize=11, fontweight='bold')
ax1.set_title('Ventas por Producto', fontsize=12, fontweight='bold', pad=10)
ax1.grid(True, axis='y', alpha=0.3, linestyle='--')
ax1.tick_params(axis='x', rotation=45)

# ========== GR√ÅFICO 2: Ingresos vs Ganancias ==========
x_pos = np.arange(len(df_productos))
width = 0.35

barras_ingresos = ax2.bar(x_pos - width/2, df_productos['ingresos'], width,
                          label='Ingresos', color=colores['primary'], alpha=0.85, edgecolor='black', linewidth=1)
barras_ganancias = ax2.bar(x_pos + width/2, df_productos['ganancia'], width,
                           label='Ganancias', color=colores['accent'], alpha=0.85, edgecolor='black', linewidth=1)

# ‚úÖ A√±adir valores en barras
for barras in [barras_ingresos, barras_ganancias]:
    for bar in barras:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                 f'${int(height):,}',
                 ha='center', va='bottom', fontsize=8, fontweight='bold')

# ‚úÖ L√≠nea de referencia
promedio_ganancia = df_productos['ganancia'].mean()
ax2.axhline(y=promedio_ganancia, color='red', linestyle='--', linewidth=2, alpha=0.6, label='Ganancia Promedio')
ax2.text(4.5, promedio_ganancia + 1000, f'Promedio: ${promedio_ganancia:,.0f}',
         ha='right', fontsize=9, fontweight='bold', 
         bbox=dict(boxstyle='round', facecolor='#ff9999', alpha=0.5))

ax2.set_ylabel('Monto (USD)', fontsize=11, fontweight='bold')
ax2.set_title('An√°lisis Financiero por Producto', fontsize=12, fontweight='bold', pad=10)
ax2.set_xticks(x_pos)
ax2.set_xticklabels(df_productos.index)
ax2.tick_params(axis='x', rotation=45)
ax2.legend(loc='upper left', fontsize=10, framealpha=0.95)
ax2.grid(True, axis='y', alpha=0.3, linestyle='--')

# ‚úÖ T√≠tulo general y contexto
fig.suptitle('üìä Dashboard: An√°lisis de Ventas Online - Q1 2024', 
             fontsize=15, fontweight='bold', y=0.98)

plt.figtext(0.5, 0.01, 'Datos simulados | Proyecto de Data Science | Alura ONE G9',
            ha='center', fontsize=9, style='italic', color='gray')

plt.tight_layout(rect=[0, 0.02, 1, 0.97])
plt.show()

print("=" * 70)
print("‚úÖ DASHBOARD PROFESIONAL COMPLETADO")
print("=" * 70)
print("\nElementos utilizados:")
print("  1. xlabel() y ylabel() - Labels descriptivos con unidades")
print("  2. xticks() y yticks() - Rotaci√≥n y formato personalizado")
print("  3. text() - Valores sobre barras y l√≠nea de referencia")
print("  4. annotate() - Se√±al con flecha al producto m√°s vendido")
print("  5. title() - T√≠tulo principal y subt√≠tulos")
print("  6. legend() - Leyenda posicionada estrat√©gicamente")
print("  7. grid() - Cuadr√≠cula sutil para mejor legibilidad")
print("  8. axhline() - L√≠nea de referencia para comparaci√≥n")
print("  9. Paleta de colores coherente y profesional")
print("\n‚úÖ Este es un ejemplo de visualizaci√≥n lista para presentaci√≥n")
print("=" * 70)
```
